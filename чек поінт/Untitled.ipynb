```python
# ğŸ§  Ğ¦ĞµĞ¹ Ğ±Ğ»Ğ¾Ğº ĞºĞ¾Ğ´Ñƒ Ğ·Ğ°Ğ²Ğ°Ğ½Ñ‚Ğ°Ğ¶ÑƒÑ” Ğ´Ğ°Ñ‚Ğ°ÑĞµÑ‚ MNIST

import tensorflow as tf


# Ğ—Ğ°Ğ²Ğ°Ğ½Ñ‚Ğ°Ğ¶ĞµĞ½Ğ½Ñ Ğ½Ğ°Ğ±Ğ¾Ñ€Ñƒ Ğ·Ğ¾Ğ±Ñ€Ğ°Ğ¶ĞµĞ½ÑŒ Ñ€ÑƒĞºĞ¾Ğ¿Ğ¸ÑĞ½Ğ¸Ñ… Ñ†Ğ¸Ñ„Ñ€
from tensorflow.keras.datasets import mnist
(x_train, y_train), (x_test, y_test) = mnist.load_data()


# Ğ’Ğ¸Ğ²ĞµĞ´ĞµĞ½Ğ½Ñ Ñ€Ğ¾Ğ·Ğ¼Ñ–Ñ€Ñ–Ğ² Ğ²Ğ¸Ğ±Ñ–Ñ€Ğ¾Ğº | Ğ¡ĞºÑ–Ğ»ÑŒĞºĞ¸ ĞºĞ°Ñ€Ñ‚Ğ¸Ğ½Ğ¾Ğº Ñ” | Ñ€Ğ¾Ğ·Ğ¼Ñ–Ñ€ (28Ã—28 Ğ¿Ñ–ĞºÑĞµĞ»Ñ–Ğ²) | Ğ¡ĞºÑ–Ğ»ÑŒĞºĞ¸ Ğ²Ñ–Ğ´Ğ¿Ğ¾Ğ²Ñ–Ğ´Ğ½Ğ¸Ñ… Ğ¼Ñ–Ñ‚Ğ¾Ğº (ÑĞºĞ° Ğ²Ñ–Ğ´Ğ¿Ğ¾Ğ²Ñ–Ğ´ÑŒ Ğ¿Ñ€Ğ°Ğ²Ğ¸Ğ»ÑŒĞ½Ğ°)
print(f"Train images shape: {x_train.shape}, Train labels shape: {y_train.shape}")
print(f"Test images shape: {x_test.shape}, Test labels shape: {y_test.shape}")
```

```python
# ğŸ§  Ğ¦ĞµĞ¹ Ğ±Ğ»Ğ¾Ğº ĞºĞ¾Ğ´Ñƒ Ğ³Ğ¾Ñ‚ÑƒÑ” Ğ·Ğ¾Ğ±Ñ€Ğ°Ğ¶ĞµĞ½Ğ½Ñ Ğ· Ñ†Ğ¸Ñ„Ñ€Ğ°Ğ¼Ğ¸ Ğ²Ñ–Ğ´ 2 Ğ´Ğ¾ 9

import numpy as np

# Ğ—Ğ°Ğ²Ğ°Ğ½Ñ‚Ğ°Ğ¶ĞµĞ½Ğ½Ñ MNIST
(x_train, y_train), (x_test, y_test) = mnist.load_data()


# Ğ¤Ñ–Ğ»ÑŒÑ‚Ñ€Ğ°Ñ†Ñ–Ñ Ñ†Ğ¸Ñ„Ñ€ Ğ²Ñ–Ğ´ 2 Ğ´Ğ¾ 9
mask_train = (y_train >= 2) & (y_train <= 9)
mask_test = (y_test >= 2) & (y_test <= 9)


# Ñ„Ñ–Ğ»ÑŒÑ‚Ñ€Ğ°Ñ†Ñ–Ñ Ğ´Ğ°Ñ‚Ğ°ÑĞµÑ‚Ñƒ, Ñ‰Ğ¾Ğ± Ğ·Ğ°Ğ»Ğ¸ÑˆĞ¸Ñ‚Ğ¸ Ñ‚Ñ–Ğ»ÑŒĞºĞ¸ Ğ¿Ğ¾Ñ‚Ñ€Ñ–Ğ±Ğ½Ñ– Ñ†Ğ¸Ñ„Ñ€Ğ¸ (2â€“9)
x_train_filtered, y_train_filtered = x_train[mask_train], y_train[mask_train]
x_test_filtered, y_test_filtered = x_test[mask_test], y_test[mask_test]

print(f"Ğ’Ñ–Ğ´Ñ„Ñ–Ğ»ÑŒÑ‚Ñ€Ğ¾Ğ²Ğ°Ğ½Ñ– Ñ‚Ñ€ĞµĞ½ÑƒĞ²Ğ°Ğ»ÑŒĞ½Ñ– Ğ·Ğ¾Ğ±Ñ€Ğ°Ğ¶ĞµĞ½Ğ½Ñ: {x_train_filtered.shape}")
print(f"Ğ’Ñ–Ğ´Ñ„Ñ–Ğ»ÑŒÑ‚Ñ€Ğ¾Ğ²Ğ°Ğ½Ñ– Ñ‚ĞµÑÑ‚Ğ¾Ğ²Ñ– Ğ·Ğ¾Ğ±Ñ€Ğ°Ğ¶ĞµĞ½Ğ½Ñ: {x_test_filtered.shape}")


# ĞĞ¾Ñ€Ğ¼Ğ°Ğ»Ñ–Ğ·Ğ°Ñ†Ñ–Ñ Ğ¿Ñ–ĞºÑĞµĞ»ÑŒĞ½Ğ¸Ñ… Ğ·Ğ½Ğ°Ñ‡ĞµĞ½ÑŒ, Ğ¿Ñ€Ğ¸Ğ±Ğ¸Ñ€Ğ°Ğ½Ğ½Ñ Ğ³Ñ€Ğ°Ğ´Ñ–Ñ”Ğ½Ñ‚Ñƒ (Ğ¿ĞµÑ€ĞµĞ²ĞµĞ´ĞµĞ½Ğ½Ñ Ğ´Ñ–Ğ°Ğ¿Ğ°Ğ·Ğ¾Ğ½Ñƒ Ğ·Ğ½Ğ°Ñ‡ĞµĞ½ÑŒ [0, 255] Ñƒ [0, 1])
x_train_filtered = x_train_filtered / 255.0
x_test_filtered = x_test_filtered / 255.0


# Ğ·Ğ¼Ñ–Ğ½ÑÑ” Ñ„Ğ¾Ñ€Ğ¼Ñƒ Ğ· (ĞºÑ–Ğ»ÑŒĞºÑ–ÑÑ‚ÑŒ, 28, 28) Ğ½Ğ° (ĞºÑ–Ğ»ÑŒĞºÑ–ÑÑ‚ÑŒ, 28, 28, 1), Ğ´Ğµ Ğ¾Ğ´Ğ¸Ğ½ Ñ†Ğµ Ğ¾Ğ´Ğ¸Ğ½ ĞºĞ¾Ğ»Ñ–Ñ€
x_train_filtered = np.expand_dims(x_train_filtered, axis=-1)  # (ĞºÑ–Ğ»ÑŒĞºÑ–ÑÑ‚ÑŒ, 28, 28, 1)
x_test_filtered = np.expand_dims(x_test_filtered, axis=-1)

print(f"Ğ¤Ğ¾Ñ€Ğ¼Ğ° Ğ¿Ñ–ÑĞ»Ñ Ñ€Ğ¾Ğ·ÑˆĞ¸Ñ€ĞµĞ½Ğ½Ñ: {x_train_filtered.shape}, {x_test_filtered.shape}")
```

```python
# ğŸ§  9 Ğ²Ğ¸Ğ¿Ğ°Ğ´ĞºĞ¾Ğ²Ğ¸Ñ… Ğ·Ğ¾Ğ±Ñ€Ğ°Ğ¶ĞµĞ½ÑŒ Ñ†Ğ¸Ñ„Ñ€ Ñ–Ğ· Ğ²Ñ–Ğ´Ñ„Ñ–Ğ»ÑŒÑ‚Ñ€Ğ¾Ğ²Ğ°Ğ½Ğ¾Ğ³Ğ¾ Ğ½Ğ°Ğ±Ğ¾Ñ€Ñƒ

import matplotlib.pyplot as plt
import numpy as np

# Ğ’Ğ¸Ğ¿Ğ°Ğ´ĞºĞ¾Ğ²Ğ¸Ğ¹ Ğ²Ğ¸Ğ±Ñ–Ñ€ Ğ·Ğ¾Ğ±Ñ€Ğ°Ğ¶ĞµĞ½ÑŒ
num_images = 9  # ĞšÑ–Ğ»ÑŒĞºÑ–ÑÑ‚ÑŒ Ğ·Ğ¾Ğ±Ñ€Ğ°Ğ¶ĞµĞ½ÑŒ Ğ´Ğ»Ñ Ğ¿Ğ¾ĞºĞ°Ğ·Ñƒ
indices = np.random.randint(0, x_train_filtered.shape[0], num_images)  # Ğ’Ğ¸Ğ¿Ğ°Ğ´ĞºĞ¾Ğ²Ñ– Ñ–Ğ½Ğ´ĞµĞºÑĞ¸


# ĞŸĞ¾Ğ±ÑƒĞ´Ğ¾Ğ²Ğ° 3x3 ÑÑ–Ñ‚ĞºĞ¸ Ğ·Ğ¾Ğ±Ñ€Ğ°Ğ¶ĞµĞ½ÑŒ
plt.figure(figsize=(8, 8))
for i, idx in enumerate(indices):
    plt.subplot(3, 3, i + 1)  # 3x3 ÑÑ–Ñ‚ĞºĞ°
    plt.imshow(x_train_filtered[idx].reshape(28, 28), cmap='gray')  # ĞŸĞ¾ĞºĞ°Ğ· Ğ·Ğ¾Ğ±Ñ€Ğ°Ğ¶ĞµĞ½Ğ½Ñ
    plt.title(f"Label: {y_train_filtered[idx]}") # ĞĞ°Ğ·Ğ²Ğ° Ğ· Ğ¼Ñ–Ñ‚ĞºĞ¾Ñ
    plt.axis('off')  # Ğ¥Ğ¾Ğ²Ğ°Ñ” Ğ¾ÑÑ– ĞºĞ¾Ğ¾Ñ€Ğ´Ğ¸Ğ½Ğ°Ñ‚

plt.suptitle("ĞŸÑ€Ğ¸ĞºĞ»Ğ°Ğ´Ğ¸ Ğ·Ğ¾Ğ±Ñ€Ğ°Ğ¶ĞµĞ½ÑŒ Ğ· Ğ²Ñ–Ğ´Ñ„Ñ–Ğ»ÑŒÑ‚Ñ€Ğ¾Ğ²Ğ°Ğ½Ğ¾Ğ³Ğ¾ Ğ½Ğ°Ğ±Ğ¾Ñ€Ñƒ")
plt.show()
```

```python
# ğŸ§  Ğ“Ğ¾Ñ‚ÑƒĞ²Ğ°Ğ½Ğ½Ñ Ğ´Ğ°Ñ‚Ğ°ÑĞµÑ‚Ñƒ MNIST (Ñ†Ğ¸Ñ„Ñ€Ğ¸) Ğ´Ğ¾ Ğ½Ğ°Ğ²Ñ‡Ğ°Ğ½Ğ½Ñ Ğ½ĞµĞ¹Ñ€Ğ¾Ğ½Ğ½Ğ¾Ñ— Ğ¼ĞµÑ€ĞµĞ¶Ñ–

import numpy as np
from tensorflow.keras.datasets import mnist
from tensorflow.keras.utils import to_categorical

# Ğ—Ğ°Ğ²Ğ°Ğ½Ñ‚Ğ°Ğ¶ĞµĞ½Ğ½Ñ Ğ´Ğ°Ğ½Ğ¸Ñ… Ğ´Ğ°Ñ‚Ğ°ÑĞµÑ‚Ñƒ
(x_train, y_train), (x_test, y_test) = mnist.load_data()


# Ğ¤Ñ–Ğ»ÑŒÑ‚Ñ€Ğ°Ñ†Ñ–Ñ, Ñ‚Ñ–Ğ»ÑŒĞºĞ¸ Ñ†Ğ¸Ñ„Ñ€Ğ¸ 2 - 9
mask_train = (y_train >= 2) & (y_train <= 9)
mask_test = (y_test >= 2) & (y_test <= 9)

x_train_filtered, y_train_filtered = x_train[mask_train], y_train[mask_train]
x_test_filtered, y_test_filtered = x_test[mask_test], y_test[mask_test]

print(f"Ğ’Ñ–Ğ´Ñ„Ñ–Ğ»ÑŒÑ‚Ñ€Ğ¾Ğ²Ğ°Ğ½Ñ– Ñ‚Ñ€ĞµĞ½ÑƒĞ²Ğ°Ğ»ÑŒĞ½Ñ– Ğ·Ğ¾Ğ±Ñ€Ğ°Ğ¶ĞµĞ½Ğ½Ñ: {x_train_filtered.shape}")
print(f"Ğ’Ñ–Ğ´Ñ„Ñ–Ğ»ÑŒÑ‚Ñ€Ğ¾Ğ²Ğ°Ğ½Ñ– Ñ‚ĞµÑÑ‚Ğ¾Ğ²Ñ– Ğ·Ğ¾Ğ±Ñ€Ğ°Ğ¶ĞµĞ½Ğ½Ñ: {x_test_filtered.shape}")


# ĞĞ¾Ñ€Ğ¼Ğ°Ğ»Ñ–Ğ·Ğ°Ñ†Ñ–Ñ Ğ´Ğ°Ğ½Ğ¸Ñ… (Ğ¿ĞµÑ€ĞµĞ²ĞµĞ´ĞµĞ½Ğ½Ñ Ğ´Ñ–Ğ°Ğ¿Ğ°Ğ·Ğ¾Ğ½Ñƒ Ğ·Ğ½Ğ°Ñ‡ĞµĞ½ÑŒ [0, 255] Ñƒ [0, 1])
x_train_filtered = x_train_filtered.astype('float32') / 255.0
x_test_filtered = x_test_filtered.astype('float32') / 255.0

# Ğ—Ğ¼Ñ–Ğ½Ğ° Ñ„Ğ¾Ñ€Ğ¼Ğ¸ Ğ´Ğ»Ñ Ğ¿Ñ–Ğ´Ğ³Ğ¾Ñ‚Ğ¾Ğ²ĞºĞ¸ Ğ´Ğ¾ Ğ½Ğ°Ğ²Ñ‡Ğ°Ğ½Ğ½Ñ (Ğ´Ğ¾Ğ´Ğ°Ğ²Ğ°Ğ½Ğ½Ñ Ğ¾ÑÑ– ĞºĞ°Ğ½Ğ°Ğ»Ñ–Ğ²)
x_train_filtered = np.expand_dims(x_train_filtered, axis=-1)
x_test_filtered = np.expand_dims(x_test_filtered, axis=-1)

print(f"Ğ¤Ğ¾Ñ€Ğ¼Ğ° Ğ¿Ñ–ÑĞ»Ñ Ñ€Ğ¾Ğ·ÑˆĞ¸Ñ€ĞµĞ½Ğ½Ñ: {x_train_filtered.shape}, {x_test_filtered.shape}")


# ĞšĞ¾Ñ€Ğ¸Ğ³ÑƒĞ²Ğ°Ğ½Ğ½Ñ Ğ¼Ñ–Ñ‚Ğ¾Ğº Ğ´Ğ»Ñ Ğ²Ñ–Ğ´Ğ¿Ğ¾Ğ²Ñ–Ğ´Ğ½Ğ¾ÑÑ‚Ñ– Ğ´Ñ–Ğ°Ğ¿Ğ°Ğ·Ğ¾Ğ½Ñƒ [0, 7]
y_train_filtered = y_train_filtered - 2
y_test_filtered = y_test_filtered - 2


# ĞŸĞµÑ€ĞµÑ‚Ğ²Ğ¾Ñ€ĞµĞ½Ğ½Ñ Ğ¼Ñ–Ñ‚Ğ¾Ğº Ñƒ Ñ„Ğ¾Ñ€Ğ¼Ğ°Ñ‚ one-hot encoding Ğ´Ğ»Ñ Ğ½ĞµĞ¹Ñ€Ğ¾Ğ½Ğ½Ğ¾Ñ— Ğ¼ĞµÑ€ĞµĞ¶Ñ–
y_train_filtered = to_categorical(y_train_filtered, num_classes=8)
y_test_filtered = to_categorical(y_test_filtered, num_classes=8)

print(f"Ğ¤Ğ¾Ñ€Ğ¼Ğ° Ğ¼Ñ–Ñ‚Ğ¾Ğº Ğ¿Ñ–ÑĞ»Ñ Ğ¿ĞµÑ€ĞµÑ‚Ğ²Ğ¾Ñ€ĞµĞ½Ğ½Ñ: {y_train_filtered.shape}, {y_test_filtered.shape}")
```

```python
# ğŸ§  Ğ†Ğ¼Ğ¿Ğ¾Ñ€Ñ‚Ğ¸ Ğ´Ğ»Ñ: Ğ¿Ğ¾Ğ±ÑƒĞ´Ğ¾Ğ²Ğ¸ Ğ°Ñ€Ñ…Ñ–Ñ‚ĞµĞºÑ‚ÑƒÑ€Ğ¸ Ğ¼Ğ¾Ğ´ĞµĞ»Ñ–, Ğ´Ğ¾Ğ´Ğ°Ğ²Ğ°Ğ½Ğ½Ñ ÑˆĞ°Ñ€Ñ–Ğ², ÑÑ‚Ğ²Ğ¾Ñ€ĞµĞ½Ğ½Ñ Ğ¼Ğ¾Ğ´ĞµĞ»Ñ–

import tensorflow as tf
from tensorflow.keras import layers, models
```

```python
# ğŸ§  Ğ¡Ñ‚Ğ²Ğ¾Ñ€ĞµĞ½ÑÑ Ñ‚Ğ° Ğ²Ğ¸Ğ²Ğ¾Ğ´Ğ¶ĞµĞ½Ğ½Ñ Ğ°Ñ€Ñ…Ñ–Ñ‚ĞµĞºÑ‚ÑƒÑ€Ğ¸ Ğ½ĞµĞ¹Ñ€Ğ¾Ğ½Ğ½Ğ¾Ñ— Ğ¼ĞµÑ€ĞµĞ¶Ñ– :3 

def create_cnn_model():
    # ĞœĞ¾Ğ´ĞµĞ»ÑŒ ÑÑ‚Ğ²Ğ¾Ñ€ÑÑ”Ñ‚ÑŒÑÑ ÑĞº Ğ¿Ğ¾ÑĞ»Ñ–Ğ´Ğ¾Ğ²Ğ½Ñ–ÑÑ‚ÑŒ ÑˆĞ°Ñ€Ñ–Ğ²
    model = models.Sequential([
        layers.Input(shape=(28, 28, 1)),  # Ğ—Ğ°Ğ´Ğ°Ğ½Ğ½Ñ Ñ„Ğ¾Ñ€Ğ¼Ğ¸ Ğ²Ñ…Ñ–Ğ´Ğ½Ğ¾Ğ³Ğ¾ Ğ·Ğ¾Ğ±Ñ€Ğ°Ğ¶ĞµĞ½Ğ½Ñ
        
        # ĞŸĞµÑ€ÑˆĞ¸Ğ¹ Ğ·Ğ³Ğ¾Ñ€Ñ‚ĞºĞ¾Ğ²Ğ¸Ğ¹ ÑˆĞ°Ñ€, 32 Ñ„Ñ–Ğ»ÑŒÑ‚Ñ€Ğ¸ Ñ€Ğ¾Ğ·Ğ¼Ñ–Ñ€Ğ¾Ğ¼ 3x3, Ğ°ĞºÑ‚Ğ¸Ğ²Ğ°Ñ†Ñ–Ñ ReLU.
        layers.Conv2D(32, (3, 3), activation='relu'),
        layers.MaxPooling2D((2, 2)),

        # Ğ”Ñ€ÑƒĞ³Ğ¸Ğ¹ Ğ·Ğ³Ğ¾Ñ€Ñ‚ĞºĞ¾Ğ²Ğ¸Ğ¹ ÑˆĞ°Ñ€, 64 Ñ„Ñ–Ğ»ÑŒÑ‚Ñ€Ğ¸, Ñ‰Ğµ Ğ¾Ğ´Ğ¸Ğ½ Ğ¿ÑƒĞ»Ñ–Ğ½Ğ³
        layers.Conv2D(64, (3, 3), activation='relu'),
        layers.MaxPooling2D((2, 2)),

        # Ğ¢Ñ€ĞµÑ‚Ñ–Ğ¹ Ğ·Ğ³Ğ¾Ñ€Ñ‚ĞºĞ¾Ğ²Ğ¸Ğ¹ ÑˆĞ°Ñ€, Ñ‰Ğµ 64 Ñ„Ñ–Ğ»ÑŒÑ‚Ñ€Ğ¸, Ğ±ĞµĞ· Ğ¿ÑƒĞ»Ñ–Ğ½Ğ³Ñƒ
        layers.Conv2D(64, (3, 3), activation='relu'),

        # ĞŸĞµÑ€ĞµÑ‚Ğ²Ğ¾Ñ€ĞµĞ½Ğ½Ñ Ñƒ Ğ²ĞµĞºÑ‚Ğ¾Ñ€
        layers.Flatten(),

        # ĞŸĞ¾Ğ²Ğ½Ğ¾Ğ·Ğ²â€™ÑĞ·Ğ½Ğ¸Ğ¹ ÑˆĞ°Ñ€
        layers.Dense(64, activation='relu'),
        layers.Dense(8, activation='softmax')  # 8 ĞºĞ»Ğ°ÑÑ–Ğ² (Ñ†Ğ¸Ñ„Ñ€Ğ¸ 2-9)
    ])
    return model

# Ğ¡Ñ‚Ğ²Ğ¾Ñ€ĞµĞ½Ğ½Ñ Ğ¼Ğ¾Ğ´ĞµĞ»Ñ–
model = create_cnn_model()

# Ğ’Ğ¸Ğ²ĞµĞ´ĞµĞ½Ğ½Ñ Ğ°Ñ€Ñ…Ñ–Ñ‚ĞµĞºÑ‚ÑƒÑ€Ğ¸ Ğ¼Ğ¾Ğ´ĞµĞ»Ñ–
model.summary()
```

```python
# ğŸ§  ĞĞ°Ğ»Ğ°ÑˆÑ‚ÑƒĞ²Ğ°Ğ½Ğ½Ñ (ĞºĞ¾Ğ¼Ğ¿Ñ–Ğ»ÑĞ²Ğ°Ğ½Ğ½Ñ) Ğ¼Ğ¾Ğ´ĞµĞ»Ñ– Ğ¿ĞµÑ€ĞµĞ´ Ñ—Ñ— Ğ½Ğ°Ğ²Ñ‡Ğ°Ğ½Ğ½ÑĞ¼

model.compile(optimizer='adam',
              loss='categorical_crossentropy',
              metrics=['accuracy'])
```

```python
# ğŸ§  ĞĞ°Ğ²Ñ‡Ğ°Ğ½Ğ½Ñ Ğ¼Ğ¾Ğ´ĞµĞ»Ñ–
history = model.fit(x_train_filtered, y_train_filtered, epochs=10, 
                    validation_data=(x_test_filtered, y_test_filtered))
```

```python
# ğŸ§  Ğ²Ñ–Ğ·ÑƒĞ°Ğ»Ñ–Ğ·Ğ°Ñ†Ñ–Ñ Ğ¿Ñ€Ğ¾Ñ†ĞµÑÑÑƒ Ğ½Ğ°Ğ²Ñ‡Ğ°Ğ½Ğ½Ñ Ğ½ĞµĞ¹Ñ€Ğ¾Ğ½Ğ½Ğ¾Ñ— Ğ¼ĞµÑ€ĞµĞ¶Ñ–

import matplotlib.pyplot as plt

# ĞŸĞ¾Ğ±ÑƒĞ´Ğ¾Ğ²Ğ° Ğ³Ñ€Ğ°Ñ„Ñ–ĞºĞ° Ğ²Ñ‚Ñ€Ğ°Ñ‚ (loss)
plt.plot(history.history['loss'], label='Ğ’Ñ‚Ñ€Ğ°Ñ‚Ğ¸ Ğ½Ğ° Ñ‚Ñ€ĞµĞ½ÑƒĞ²Ğ°Ğ½Ğ½Ñ–')
plt.plot(history.history['val_loss'], label='Ğ’Ñ‚Ñ€Ğ°Ñ‚Ğ¸ Ğ½Ğ° Ğ²Ğ°Ğ»Ñ–Ğ´Ğ°Ñ†Ñ–Ñ—')
plt.xlabel('Ğ•Ğ¿Ğ¾Ñ…Ğ¸')
plt.ylabel('Ğ’Ñ‚Ñ€Ğ°Ñ‚Ğ¸')
plt.legend()
plt.title('Ğ“Ñ€Ğ°Ñ„Ñ–Ğº Ğ²Ñ‚Ñ€Ğ°Ñ‚')
plt.show()

# ĞŸĞ¾Ğ±ÑƒĞ´Ğ¾Ğ²Ğ° Ğ³Ñ€Ğ°Ñ„Ñ–ĞºĞ° Ñ‚Ğ¾Ñ‡Ğ½Ğ¾ÑÑ‚Ñ– (accuracy)
plt.plot(history.history['accuracy'], label='Ğ¢Ğ¾Ñ‡Ğ½Ñ–ÑÑ‚ÑŒ Ğ½Ğ° Ñ‚Ñ€ĞµĞ½ÑƒĞ²Ğ°Ğ½Ğ½Ñ–')
plt.plot(history.history['val_accuracy'], label='Ğ¢Ğ¾Ñ‡Ğ½Ñ–ÑÑ‚ÑŒ Ğ½Ğ° Ğ²Ğ°Ğ»Ñ–Ğ´Ğ°Ñ†Ñ–Ñ—')
plt.xlabel('Ğ•Ğ¿Ğ¾Ñ…Ğ¸')
plt.ylabel('Ğ¢Ğ¾Ñ‡Ğ½Ñ–ÑÑ‚ÑŒ')
plt.legend()
plt.title('Ğ“Ñ€Ğ°Ñ„Ñ–Ğº Ñ‚Ğ¾Ñ‡Ğ½Ğ¾ÑÑ‚Ñ–')
plt.show()
```

```python
# ğŸ§  Ğ—Ğ±ĞµÑ€Ñ–Ğ³Ğ°Ğ½Ğ½Ñ Ğ½Ğ°Ğ²Ñ‡ĞµĞ½Ñƒ Ğ½ĞµĞ¹Ñ€Ğ¾Ğ½Ğ½Ñƒ Ğ¼ĞµÑ€ĞµĞ¶Ñƒ model Ñƒ Ñ„Ğ°Ğ¹Ğ» speed_limit_model.keras

model.save('speed_limit_model.keras')
print("ĞœĞ¾Ğ´ĞµĞ»ÑŒ ÑƒÑĞ¿Ñ–ÑˆĞ½Ğ¾ Ğ·Ğ±ĞµÑ€ĞµĞ¶ĞµĞ½Ğ° Ñƒ Ñ„Ğ¾Ñ€Ğ¼Ğ°Ñ‚Ñ– .keras")
```

```python
#  ğŸ§ ğŸ§ ğŸ§  Ğ“Ğ¾Ñ‚Ğ¾Ğ²Ğ° Ğ»Ğ°Ğ±Ğ¾Ñ€Ğ°Ñ‚Ğ¾Ñ€Ğ½Ğ° Ğ½Ğ¾Ğ¼ĞµÑ€ 444 ğŸ§ ğŸ§ ğŸ§ 


# === Ğ†ĞœĞŸĞĞ Ğ¢Ğ˜ Ğ¢Ğ Ğ†ĞĞ†Ğ¦Ğ†ĞĞ›Ğ†Ğ—ĞĞ¦Ğ†Ğ¯ ===
import os, cv2, random
import numpy as np
import networkx as nx
import matplotlib.pyplot as plt
from tensorflow.keras.models import load_model
from collections import deque
from IPython.display import display, Markdown


# === Ğ—ĞĞ“Ğ Ğ£Ğ—ĞšĞ ĞœĞĞ”Ğ•Ğ›Ğ† Ğ¢Ğ Ğ†ĞĞ†Ğ¦Ğ†ĞĞ›Ğ†Ğ—ĞĞ¦Ğ†Ğ¯ ĞŸĞĞŸĞĞš ===
#from tensorflow.keras.models import load_model
model = load_model('speed_limit_model.keras')
print("ĞœĞ¾Ğ´ĞµĞ»ÑŒ Ğ¾Ğ±Ğ¼ĞµĞ¶ĞµĞ½Ğ½Ñ ÑˆĞ²Ğ¸Ğ´ĞºĞ¾ÑÑ‚Ñ– ÑƒÑĞ¿Ñ–ÑˆĞ½Ğ¾ Ğ·Ğ°Ğ²Ğ°Ğ½Ñ‚Ğ°Ğ¶ĞµĞ½Ğ°.")


# ĞŸĞµÑ€ĞµĞ²Ñ–Ñ€ĞºĞ° Ğ½Ğ°ÑĞ²Ğ½Ğ¾ÑÑ‚Ñ– Ğ¿Ğ°Ğ¿ĞºĞ¸ Ğ·Ñ– Ğ·Ğ½Ğ°ĞºĞ°Ğ¼Ğ¸
if not os.path.exists('signs'):
    os.makedirs('signs')

road_signs = {}
# Ğ“Ğ»Ğ¾Ğ±Ğ°Ğ»ÑŒĞ½Ğ° Ğ·Ğ¼Ñ–Ğ½Ğ½Ğ° Ğ´Ğ»Ñ Ğ·Ğ±ĞµÑ€ĞµĞ¶ĞµĞ½Ğ½Ñ Ğ¾ÑÑ‚Ğ°Ğ½Ğ½ÑŒĞ¾Ğ³Ğ¾ Ğ²Ğ¸ĞºĞ¾Ñ€Ğ¸ÑÑ‚Ğ°Ğ½Ğ¾Ğ³Ğ¾ Ğ·Ğ½Ğ°ĞºĞ°
last_sign_image = None

#result = move_agent_with_knowledge(G, kb, test_start, test_end, max_steps=100)
#print(f"ĞšÑ–Ğ»ÑŒĞºÑ–ÑÑ‚ÑŒ Ğ¿Ğ¾Ğ²ĞµÑ€Ğ½ÑƒÑ‚Ğ¸Ñ… Ğ·Ğ½Ğ°Ñ‡ĞµĞ½ÑŒ: {len(result)}")





# ğŸ§©ğŸ§©ğŸ§©ğŸ§© Ğ ĞĞ‘ĞĞ¢Ğ Ğ— Ğ“Ğ ĞĞ¤ĞĞœ ğŸ§©ğŸ§©ğŸ§©ğŸ§©


def generate_graph(width, height):
    """Ğ¡Ñ‚Ğ²Ğ¾Ñ€ĞµĞ½Ğ½Ñ ÑÑ–Ñ‚ĞºĞ¾Ğ²Ğ¾Ğ³Ğ¾ Ğ³Ñ€Ğ°Ñ„Ğ° Ğ· Ğ¾Ğ±Ğ¼ĞµĞ¶ĞµĞ½Ğ½ÑĞ¼Ğ¸ ÑˆĞ²Ğ¸Ğ´ĞºĞ¾ÑÑ‚Ñ–."""
    G = nx.Graph()
    speed_limits = {}

    for i in range(width):
        for j in range(height):
            G.add_node((i, j))  # Ğ”Ğ¾Ğ´Ğ°Ğ²Ğ°Ğ½Ğ½Ñ Ğ²ÑƒĞ·Ğ»Ğ° Ğ½Ğ° ĞºĞ¾Ğ¾Ñ€Ğ´Ğ¸Ğ½Ğ°Ñ‚Ğ¸ (i, j)

            # Ğ”Ğ¾Ğ´Ğ°Ğ²Ğ°Ğ½Ğ½Ñ Ñ€ĞµĞ±Ñ€Ğ° Ğ»Ñ–Ğ²Ğ¾Ñ€ÑƒÑ‡ (ÑĞºÑ‰Ğ¾ Ğ½Ğµ Ğ½Ğ° Ğ¿ĞµÑ€ÑˆĞ¾Ğ¼Ñƒ ÑÑ‚Ğ¾Ğ²Ğ¿Ñ‡Ğ¸ĞºÑƒ)
            if i > 0:
                G.add_edge((i, j), (i - 1, j))
                speed_limits[((i, j), (i - 1, j))] = random.choice(range(20, 100, 10))  # ĞŸÑ€Ğ¸Ğ·Ğ½Ğ°Ñ‡ĞµĞ½Ğ½Ñ Ğ²Ğ¸Ğ¿Ğ°Ğ´ĞºĞ¾Ğ²Ğ¾Ñ— ÑˆĞ²Ğ¸Ğ´ĞºĞ¾ÑÑ‚Ñ–

            # Ğ”Ğ¾Ğ´Ğ°Ğ²Ğ°Ğ½Ğ½Ñ Ñ€ĞµĞ±Ñ€Ğ° Ğ·Ğ³Ğ¾Ñ€Ğ¸ (ÑĞºÑ‰Ğ¾ Ğ½Ğµ Ğ² Ğ¿ĞµÑ€ÑˆĞ¾Ğ¼Ñƒ Ñ€ÑĞ´ĞºÑƒ)
            if j > 0:
                G.add_edge((i, j), (i, j - 1))
                speed_limits[((i, j), (i, j - 1))] = random.choice(range(20, 100, 10))  # ĞŸÑ€Ğ¸Ğ·Ğ½Ğ°Ñ‡ĞµĞ½Ğ½Ñ Ğ²Ğ¸Ğ¿Ğ°Ğ´ĞºĞ¾Ğ²Ğ¾Ñ— ÑˆĞ²Ğ¸Ğ´ĞºĞ¾ÑÑ‚Ñ–

    return G, speed_limits  # ĞŸĞ¾Ğ²ĞµÑ€Ğ½ĞµĞ½Ğ½Ñ Ğ³Ñ€Ğ°Ñ„Ğ° Ñ‚Ğ° ÑĞ»Ğ¾Ğ²Ğ½Ğ¸ĞºĞ° ÑˆĞ²Ğ¸Ğ´ĞºĞ¾ÑÑ‚ĞµĞ¹





def remove_random_edges(G, remove_count):
    """Ğ’Ğ¸Ğ´Ğ°Ğ»ĞµĞ½Ğ½Ñ Ğ²Ğ¸Ğ¿Ğ°Ğ´ĞºĞ¾Ğ²Ğ¸Ñ… Ñ€ĞµĞ±ĞµÑ€ Ñ–Ğ· Ğ¿ĞµÑ€ĞµĞ²Ñ–Ñ€ĞºĞ¾Ñ Ğ½Ğ° Ğ·Ğ²'ÑĞ·Ğ½Ñ–ÑÑ‚ÑŒ."""
    edges = list(G.edges())  # ĞÑ‚Ñ€Ğ¸Ğ¼Ğ°Ğ½Ğ½Ñ ÑĞ¿Ğ¸ÑĞºÑƒ Ğ²ÑÑ–Ñ… Ñ€ĞµĞ±ĞµÑ€
    max_removable_edges = len(edges)

    # ĞĞ±Ğ¼ĞµĞ¶ĞµĞ½Ğ½Ñ ĞºÑ–Ğ»ÑŒĞºĞ¾ÑÑ‚Ñ– Ğ²Ğ¸Ğ´Ğ°Ğ»ÑÑ”Ğ¼Ğ¸Ñ… Ñ€ĞµĞ±ĞµÑ€, ÑĞºÑ‰Ğ¾ Ğ·Ğ°Ğ´Ğ°Ğ½Ğµ Ñ‡Ğ¸ÑĞ»Ğ¾ Ğ¿ĞµÑ€ĞµĞ²Ğ¸Ñ‰ÑƒÑ” Ğ´Ğ¾ÑÑ‚ÑƒĞ¿Ğ½Ğµ
    if remove_count > max_removable_edges:
        remove_count = max_removable_edges

    # Ğ’Ğ¸Ğ¿Ğ°Ğ´ĞºĞ¾Ğ²Ğ¸Ğ¹ Ğ²Ğ¸Ğ±Ñ–Ñ€ Ñ€ĞµĞ±ĞµÑ€ Ğ´Ğ»Ñ Ğ²Ğ¸Ğ´Ğ°Ğ»ĞµĞ½Ğ½Ñ
    removed_edges = random.sample(edges, remove_count)
    
    for edge in removed_edges:
        G.remove_edge(*edge)  # Ğ’Ğ¸Ğ´Ğ°Ğ»ĞµĞ½Ğ½Ñ Ñ€ĞµĞ±Ñ€Ğ°

        # ĞŸĞµÑ€ĞµĞ²Ñ–Ñ€ĞºĞ° Ğ½Ğ° Ğ·Ğ±ĞµÑ€ĞµĞ¶ĞµĞ½Ğ½Ñ Ğ·Ğ²'ÑĞ·Ğ½Ğ¾ÑÑ‚Ñ–
        if not nx.is_connected(G):
            G.add_edge(*edge)  # Ğ’Ñ–Ğ´Ğ½Ğ¾Ğ²Ğ»ĞµĞ½Ğ½Ñ Ñ€ĞµĞ±Ñ€Ğ°, ÑĞºÑ‰Ğ¾ Ğ³Ñ€Ğ°Ñ„ Ñ€Ğ¾Ğ·Ñ–Ñ€Ğ²Ğ°Ğ²ÑÑ
            
    return G  # ĞŸĞ¾Ğ²ĞµÑ€Ğ½ĞµĞ½Ğ½Ñ Ğ¾Ğ½Ğ¾Ğ²Ğ»ĞµĞ½Ğ¾Ğ³Ğ¾ Ğ³Ñ€Ğ°Ñ„Ğ°


    




# ğŸ“·ğŸ“·ğŸ“·ğŸ“· Ğ“ĞµĞ½ĞµÑ€Ğ°Ñ†Ñ–Ñ Ñ‚Ğ° Ñ€Ğ¾Ğ·Ğ¿Ñ–Ğ·Ğ½Ğ°Ğ²Ğ°Ğ½Ğ½Ñ Ğ·Ğ½Ğ°ĞºÑ–Ğ² ğŸ“·ğŸ“·ğŸ“·ğŸ“·


def generate_speed_limit_signs(G):
    """Ğ“ĞµĞ½ĞµÑ€Ğ°Ñ†Ñ–Ñ ÑƒĞ½Ñ–ĞºĞ°Ğ»ÑŒĞ½Ğ¸Ñ… Ğ·Ğ½Ğ°ĞºÑ–Ğ² Ğ¾Ğ±Ğ¼ĞµĞ¶ĞµĞ½Ğ½Ñ ÑˆĞ²Ğ¸Ğ´ĞºĞ¾ÑÑ‚Ñ– Ğ´Ğ»Ñ ĞºĞ¾Ğ¶Ğ½Ğ¾Ğ³Ğ¾ Ğ½Ğ°Ğ¿Ñ€ÑĞ¼ĞºÑƒ Ñ€ĞµĞ±Ñ€Ğ° Ğ² Ğ³Ñ€Ğ°Ñ„Ñ–."""
    global road_signs
    os.makedirs('signs', exist_ok=True)
    road_signs = {}

    all_edges = list(G.edges())
    num_edges = len(all_edges)

    # ĞŸĞµÑ€ĞµĞ²Ñ–Ñ€ĞºĞ° Ğ½Ğ° Ğ¿ĞµÑ€ĞµĞ²Ğ¸Ñ‰ĞµĞ½Ğ½Ñ Ğ¼Ğ°ĞºÑĞ¸Ğ¼Ğ°Ğ»ÑŒĞ½Ğ¾Ñ— ĞºÑ–Ğ»ÑŒĞºĞ¾ÑÑ‚Ñ– Ğ½Ğ°Ğ¿Ñ€ÑĞ¼ĞºÑ–Ğ² (100)
    if num_edges * 2 > 100:
        raise ValueError("âš  Ğ—Ğ°Ğ½Ğ°Ğ´Ñ‚Ğ¾ Ğ±Ğ°Ğ³Ğ°Ñ‚Ğ¾ Ğ½Ğ°Ğ¿Ñ€ÑĞ¼ĞºÑ–Ğ²: Ğ¿ĞµÑ€ĞµĞ²Ğ¸Ñ‰ĞµĞ½Ğ¾ 100. Ğ—Ğ±Ñ–Ğ»ÑŒÑˆ Ğ»Ñ–Ğ¼Ñ–Ñ‚ Ğ°Ğ±Ğ¾ Ğ·Ğ¼ĞµĞ½Ñˆ Ğ³Ñ€Ğ°Ñ„.")

    # Ğ¡Ñ‚Ğ²Ğ¾Ñ€ĞµĞ½Ğ½Ñ ÑĞ¿Ğ¸ÑĞºÑƒ ÑƒĞ½Ñ–ĞºĞ°Ğ»ÑŒĞ½Ğ¸Ñ… Ñ–Ğ½Ğ´ĞµĞºÑÑ–Ğ² Ğ´Ğ»Ñ Ğ½Ğ°Ğ·Ğ² Ñ„Ğ°Ğ¹Ğ»Ñ–Ğ²
    unique_indices = list(range(1, 101))
    random.shuffle(unique_indices)
    index_iter = iter(unique_indices)

    for edge in all_edges:
        node1, node2 = edge

        # Ğ“ĞµĞ½ĞµÑ€Ğ°Ñ†Ñ–Ñ Ğ·Ğ½Ğ°ĞºÑ–Ğ² Ğ´Ğ»Ñ Ğ¾Ğ±Ğ¾Ñ… Ğ½Ğ°Ğ¿Ñ€ÑĞ¼ĞºÑ–Ğ²
        for direction in [(node1, node2), (node2, node1)]:
            idx = next(index_iter)
            sign_path = f'signs/speed_sign_{idx}.png'

            # Ğ“ĞµĞ½ĞµÑ€Ğ°Ñ†Ñ–Ñ Ğ·Ğ¾Ğ±Ñ€Ğ°Ğ¶ĞµĞ½Ğ½Ñ Ñ‚Ñ–Ğ»ÑŒĞºĞ¸ ÑĞºÑ‰Ğ¾ Ğ²Ğ¾Ğ½Ğ¾ Ñ‰Ğµ Ğ½Ğµ Ñ–ÑĞ½ÑƒÑ”
            """if not os.path.exists(sign_path):
                digit = random.randint(2, 9)
                pixel_value = digit * 25
                sign_image = np.full((28, 28), pixel_value, dtype=np.uint8)
                cv2.imwrite(sign_path, sign_image)
                print(f"ğŸ–¼ Ğ—Ğ±ĞµÑ€ĞµĞ¶ĞµĞ½Ğ¾ Ğ½Ğ¾Ğ²Ğµ Ğ·Ğ¾Ğ±Ñ€Ğ°Ğ¶ĞµĞ½Ğ½Ñ: {sign_path}") 
            else:
                print(f"ğŸ›‘ ĞŸÑ€Ğ¾Ğ¿ÑƒÑ‰ĞµĞ½Ğ¾ Ğ¿ĞµÑ€ĞµĞ·Ğ°Ğ¿Ğ¸Ñ: {sign_path}") """

            # ĞŸÑ€Ğ¸Ğ²'ÑĞ·ĞºĞ° ÑˆĞ»ÑÑ…Ñƒ Ğ·Ğ¾Ğ±Ñ€Ğ°Ğ¶ĞµĞ½Ğ½Ñ Ğ´Ğ¾ Ğ½Ğ°Ğ¿Ñ€ÑĞ¼ĞºÑƒ
            road_signs[direction] = sign_path

    print(f"âœ… Ğ¡Ñ‚Ğ²Ğ¾Ñ€ĞµĞ½Ğ¾ {len(road_signs)} Ğ·Ğ½Ğ°ĞºÑ–Ğ² (Ğ°Ğ±Ğ¾ Ğ¿Ñ€Ğ¸Ğ²'ÑĞ·Ğ¾Ğº) Ğ´Ğ»Ñ Ğ½Ğ°Ğ¿Ñ€ÑĞ¼ĞºÑ–Ğ² Ñ€ĞµĞ±ĞµÑ€.")
    print("ğŸ‰ Ğ“ĞµĞ½ĞµÑ€Ğ°Ñ†Ñ–Ñ Ğ·Ğ½Ğ°ĞºÑ–Ğ² Ğ·Ğ°Ğ²ĞµÑ€ÑˆĞµĞ½Ğ°!")





def check_signs_directory():
    """ĞŸĞµÑ€ĞµĞ²Ñ–Ñ€ĞºĞ° Ğ²Ğ¼Ñ–ÑÑ‚Ñƒ Ğ¿Ğ°Ğ¿ĞºĞ¸ 'signs' Ñ‚Ğ° Ğ³ĞµĞ½ĞµÑ€Ğ°Ñ†Ñ–Ñ Ğ·Ğ½Ğ°ĞºÑ–Ğ² Ğ·Ğ° Ğ¿Ğ¾Ñ‚Ñ€ĞµĞ±Ğ¸."""
    sign_files = os.listdir('signs')

    # Ğ¯ĞºÑ‰Ğ¾ Ğ¿Ğ°Ğ¿ĞºĞ° Ğ¿Ğ¾Ñ€Ğ¾Ğ¶Ğ½Ñ â€” Ğ²Ğ¸ĞºĞ¾Ğ½ÑƒÑ”Ñ‚ÑŒÑÑ Ğ³ĞµĞ½ĞµÑ€Ğ°Ñ†Ñ–Ñ Ğ·Ğ½Ğ°ĞºÑ–Ğ²
    if not sign_files:
        print("âš  ĞŸĞ°Ğ¿ĞºĞ° 'signs' Ğ¿Ğ¾Ñ€Ğ¾Ğ¶Ğ½Ñ!")
        generate_speed_limit_signs()
    else:
        # Ğ’Ğ¸Ğ²ĞµĞ´ĞµĞ½Ğ½Ñ ĞºÑ–Ğ»ÑŒĞºĞ¾ÑÑ‚Ñ– Ğ·Ğ½Ğ°Ğ¹Ğ´ĞµĞ½Ğ¸Ñ… Ñ„Ğ°Ğ¹Ğ»Ñ–Ğ²
        print(f"âœ… Ğ£ Ğ¿Ğ°Ğ¿Ñ†Ñ– 'signs' Ğ·Ğ½Ğ°Ğ¹Ğ´ĞµĞ½Ğ¾ {len(sign_files)} Ñ„Ğ°Ğ¹Ğ»Ñ–Ğ².")

check_signs_directory()





def recognize_speed_limit(sign_image_path, model):
    """Ğ Ğ¾Ğ·Ğ¿Ñ–Ğ·Ğ½Ğ°Ğ²Ğ°Ğ½Ğ½Ñ Ğ·Ğ½Ğ°ĞºĞ° ÑˆĞ²Ğ¸Ğ´ĞºĞ¾ÑÑ‚Ñ– Ğ·Ğ° Ğ´Ğ¾Ğ¿Ğ¾Ğ¼Ğ¾Ğ³Ğ¾Ñ Ğ¼Ğ¾Ğ´ĞµĞ»Ñ– Ğ½ĞµĞ¹Ñ€Ğ¾Ğ½Ğ½Ğ¾Ñ— Ğ¼ĞµÑ€ĞµĞ¶Ñ–."""

    # ĞŸĞµÑ€ĞµĞ²Ñ–Ñ€ĞºĞ° Ñ–ÑĞ½ÑƒĞ²Ğ°Ğ½Ğ½Ñ Ñ„Ğ°Ğ¹Ğ»Ñƒ
    if not os.path.exists(sign_image_path):
        print(f"âš  Ğ¤Ğ°Ğ¹Ğ» {sign_image_path} Ğ½Ğµ Ğ·Ğ½Ğ°Ğ¹Ğ´ĞµĞ½Ğ¾. Ğ’Ğ¸ĞºĞ¾Ñ€Ğ¸ÑÑ‚Ğ¾Ğ²ÑƒÑ”Ñ‚ÑŒÑÑ Ğ·Ğ½Ğ°Ñ‡ĞµĞ½Ğ½Ñ Ğ·Ğ° Ğ·Ğ°Ğ¼Ğ¾Ğ²Ñ‡ÑƒĞ²Ğ°Ğ½Ğ½ÑĞ¼.")
        return 50  # Ğ—Ğ½Ğ°Ñ‡ĞµĞ½Ğ½Ñ Ğ·Ğ° Ğ·Ğ°Ğ¼Ğ¾Ğ²Ñ‡ÑƒĞ²Ğ°Ğ½Ğ½ÑĞ¼

    # Ğ—Ğ°Ğ²Ğ°Ğ½Ñ‚Ğ°Ğ¶ĞµĞ½Ğ½Ñ Ğ·Ğ¾Ğ±Ñ€Ğ°Ğ¶ĞµĞ½Ğ½Ñ Ğ² Ğ³Ñ€Ğ°Ğ´Ğ°Ñ†Ñ–ÑÑ… ÑÑ–Ñ€Ğ¾Ğ³Ğ¾
    image = cv2.imread(sign_image_path, cv2.IMREAD_GRAYSCALE)
    if image is None or image.size == 0:
        print(f"âš  ĞŸĞ¾Ğ¼Ğ¸Ğ»ĞºĞ° Ğ·Ñ‡Ğ¸Ñ‚ÑƒĞ²Ğ°Ğ½Ğ½Ñ {sign_image_path}. Ğ’Ğ¸ĞºĞ¾Ñ€Ğ¸ÑÑ‚Ğ¾Ğ²ÑƒÑ”Ñ‚ÑŒÑÑ Ğ·Ğ½Ğ°Ñ‡ĞµĞ½Ğ½Ñ Ğ·Ğ° Ğ·Ğ°Ğ¼Ğ¾Ğ²Ñ‡ÑƒĞ²Ğ°Ğ½Ğ½ÑĞ¼.")
        return 50  # Ğ—Ğ½Ğ°Ñ‡ĞµĞ½Ğ½Ñ Ğ·Ğ° Ğ·Ğ°Ğ¼Ğ¾Ğ²Ñ‡ÑƒĞ²Ğ°Ğ½Ğ½ÑĞ¼

    # ĞŸÑ–Ğ´Ğ³Ğ¾Ñ‚Ğ¾Ğ²ĞºĞ° Ğ·Ğ¾Ğ±Ñ€Ğ°Ğ¶ĞµĞ½Ğ½Ñ Ğ´Ğ¾ Ğ¿ĞµÑ€ĞµĞ´Ğ±Ğ°Ñ‡ĞµĞ½Ğ½Ñ
    image = cv2.resize(image, (28, 28))               # Ğ—Ğ¼Ñ–Ğ½Ğ° Ñ€Ğ¾Ğ·Ğ¼Ñ–Ñ€Ñƒ
    image = image.astype('float32') / 255.0           # ĞĞ¾Ñ€Ğ¼Ğ°Ğ»Ñ–Ğ·Ğ°Ñ†Ñ–Ñ Ğ·Ğ½Ğ°Ñ‡ĞµĞ½ÑŒ
    image = np.expand_dims(image, axis=(0, -1))       # Ğ”Ğ¾Ğ´Ğ°Ğ²Ğ°Ğ½Ğ½Ñ Ñ€Ğ¾Ğ·Ğ¼Ñ–Ñ€Ğ½Ğ¾ÑÑ‚ĞµĞ¹

    # ĞŸĞµÑ€ĞµĞ´Ğ±Ğ°Ñ‡ĞµĞ½Ğ½Ñ Ñ†Ğ¸Ñ„Ñ€Ğ¸ Ğ½Ğ° Ğ·Ğ½Ğ°ĞºÑƒ
    prediction = model.predict(image)
    predicted_label = np.argmax(prediction) + 2       # Ğ¦Ğ¸Ñ„Ñ€Ğ¸ Ğ²Ñ–Ğ´ 2 Ğ´Ğ¾ 9
    speed_limit = predicted_label * 10                # ĞŸĞµÑ€ĞµĞ²ĞµĞ´ĞµĞ½Ğ½Ñ Ğ² ĞºĞ¼/Ğ³Ğ¾Ğ´

    print(f"ğŸ Ğ Ğ¾Ğ·Ğ¿Ñ–Ğ·Ğ½Ğ°Ğ½Ğ¾ Ğ·Ğ½Ğ°Ğº {predicted_label}, Ğ¾Ğ±Ğ¼ĞµĞ¶ĞµĞ½Ğ½Ñ: {speed_limit} ĞºĞ¼/Ğ³Ğ¾Ğ´")
    return speed_limit




"""
def predict_speed_limit(node1, node2, model):
    \"\"\"ĞŸĞµÑ€ĞµĞ´Ğ±Ğ°Ñ‡ĞµĞ½Ğ½Ñ Ğ¾Ğ±Ğ¼ĞµĞ¶ĞµĞ½Ğ½Ñ ÑˆĞ²Ğ¸Ğ´ĞºĞ¾ÑÑ‚Ñ– Ğ¼Ñ–Ğ¶ Ğ²ÑƒĞ·Ğ»Ğ°Ğ¼Ğ¸ Ğ·Ğ° Ğ·Ğ¾Ğ±Ñ€Ğ°Ğ¶ĞµĞ½Ğ½ÑĞ¼ Ğ·Ğ½Ğ°ĞºĞ°.\"\"\"

    # ĞŸĞµÑ€ĞµĞ²Ñ–Ñ€ĞºĞ° Ğ½Ğ° ÑÑ‚Ğ°Ñ€Ñ‚Ğ¾Ğ²Ñƒ Ğ¿Ğ¾Ğ·Ğ¸Ñ†Ñ–Ñ Ğ±ĞµĞ· Ğ¿Ğ¾Ğ¿ĞµÑ€ĞµĞ´Ğ½ÑŒĞ¾Ğ³Ğ¾ Ğ²ÑƒĞ·Ğ»Ğ°
    if node1 is None or node2 is None:
        return 50  # Ğ—Ğ½Ğ°Ñ‡ĞµĞ½Ğ½Ñ Ğ·Ğ° Ğ·Ğ°Ğ¼Ğ¾Ğ²Ñ‡ÑƒĞ²Ğ°Ğ½Ğ½ÑĞ¼

    # Ğ¤Ğ¾Ñ€Ğ¼ÑƒĞ²Ğ°Ğ½Ğ½Ñ ĞºĞ»ÑÑ‡Ğ° Ñ€ĞµĞ±Ñ€Ğ° Ñƒ Ñ„Ñ–ĞºÑĞ¾Ğ²Ğ°Ğ½Ğ¾Ğ¼Ñƒ Ğ¿Ğ¾Ñ€ÑĞ´ĞºÑƒ
    edge_key = tuple(sorted([node1, node2]))
    image_path = f'signs/speed_sign_{amogus_1.png'

    # ĞŸĞµÑ€ĞµĞ²Ñ–Ñ€ĞºĞ° Ğ½Ğ°ÑĞ²Ğ½Ğ¾ÑÑ‚Ñ– Ğ·Ğ¾Ğ±Ñ€Ğ°Ğ¶ĞµĞ½Ğ½Ñ
    if not os.path.exists(image_path):
        print(f"âš  Ğ¤Ğ°Ğ¹Ğ» {image_path} Ğ½Ğµ Ğ·Ğ½Ğ°Ğ¹Ğ´ĞµĞ½Ğ¾. Ğ’Ğ¸ĞºĞ¾Ñ€Ğ¸ÑÑ‚Ğ¾Ğ²ÑƒÑ”Ñ‚ÑŒÑÑ Ğ·Ğ½Ğ°Ñ‡ĞµĞ½Ğ½Ñ Ğ·Ğ° Ğ·Ğ°Ğ¼Ğ¾Ğ²Ñ‡ÑƒĞ²Ğ°Ğ½Ğ½ÑĞ¼.")
        return 50

    # Ğ—Ğ°Ğ²Ğ°Ğ½Ñ‚Ğ°Ğ¶ĞµĞ½Ğ½Ñ Ğ·Ğ¾Ğ±Ñ€Ğ°Ğ¶ĞµĞ½Ğ½Ñ Ğ² Ğ³Ñ€Ğ°Ğ´Ğ°Ñ†Ñ–ÑÑ… ÑÑ–Ñ€Ğ¾Ğ³Ğ¾
    image = cv2.imread(image_path, cv2.IMREAD_GRAYSCALE)
    if image is None or image.size == 0:
        print(f"âš  ĞĞµĞ¼Ğ¾Ğ¶Ğ»Ğ¸Ğ²Ğ¾ Ğ·Ñ‡Ğ¸Ñ‚Ğ°Ñ‚Ğ¸ Ğ·Ğ¾Ğ±Ñ€Ğ°Ğ¶ĞµĞ½Ğ½Ñ {image_path}. Ğ’Ğ¸ĞºĞ¾Ñ€Ğ¸ÑÑ‚Ğ¾Ğ²ÑƒÑ”Ñ‚ÑŒÑÑ Ğ·Ğ½Ğ°Ñ‡ĞµĞ½Ğ½Ñ Ğ·Ğ° Ğ·Ğ°Ğ¼Ğ¾Ğ²Ñ‡ÑƒĞ²Ğ°Ğ½Ğ½ÑĞ¼.")
        return 50

    # ĞŸÑ–Ğ´Ğ³Ğ¾Ñ‚Ğ¾Ğ²ĞºĞ° Ğ·Ğ¾Ğ±Ñ€Ğ°Ğ¶ĞµĞ½Ğ½Ñ
    image = cv2.resize(image, (28, 28))            # Ğ—Ğ¼Ñ–Ğ½Ğ° Ñ€Ğ¾Ğ·Ğ¼Ñ–Ñ€Ñƒ
    image = cv2.bitwise_not(image)                 # Ğ†Ğ½Ğ²ĞµÑ€ÑÑ–Ñ ĞºĞ¾Ğ»ÑŒĞ¾Ñ€Ñ–Ğ²
    image = image / 255.0                          # ĞĞ¾Ñ€Ğ¼Ğ°Ğ»Ñ–Ğ·Ğ°Ñ†Ñ–Ñ
    image = np.expand_dims(image, axis=(0, -1))    # Ğ”Ğ¾Ğ´Ğ°Ğ²Ğ°Ğ½Ğ½Ñ Ğ¾ÑĞµĞ¹

    # ĞšĞ»Ğ°ÑĞ¸Ñ„Ñ–ĞºĞ°Ñ†Ñ–Ñ Ñ‚Ğ° Ğ¾Ğ±Ñ‡Ğ¸ÑĞ»ĞµĞ½Ğ½Ñ ÑˆĞ²Ğ¸Ğ´ĞºĞ¾ÑÑ‚Ñ–
    prediction = model.predict(image)
    predicted_label = np.argmax(prediction)        # Ğ—Ğ½Ğ°Ñ‡ĞµĞ½Ğ½Ñ Ğ² Ğ´Ñ–Ğ°Ğ¿Ğ°Ğ·Ğ¾Ğ½Ñ– 0â€“7
    speed = predicted_label * 10                   # Ğ¨Ğ²Ğ¸Ğ´ĞºÑ–ÑÑ‚ÑŒ (20â€“90 ĞºĞ¼/Ğ³Ğ¾Ğ´)

    return speed
"""




# ğŸ‘€ğŸ‘€ğŸ‘€ğŸ‘€ Ğ‘Ğ°Ğ·Ğ° Ğ·Ğ½Ğ°Ğ½ÑŒ Ğ°Ğ³ĞµĞ½Ñ‚Ğ° ğŸ‘€ğŸ‘€ğŸ‘€ğŸ‘€

class KnowledgeBase:
    """Ğ‘Ğ°Ğ·Ğ° Ğ·Ğ½Ğ°Ğ½ÑŒ Ğ°Ğ³ĞµĞ½Ñ‚Ğ° Ğ´Ğ»Ñ Ğ·Ğ±ĞµÑ€Ñ–Ğ³Ğ°Ğ½Ğ½Ñ Ñ–Ğ½Ñ„Ğ¾Ñ€Ğ¼Ğ°Ñ†Ñ–Ñ— Ğ¿Ñ€Ğ¾ Ğ³Ñ€Ğ°Ñ„-Ğ´Ğ¾Ñ€Ğ¾Ğ³Ñƒ."""

    def __init__(self):
        self.visited_nodes = set()       # Ğ’ÑƒĞ·Ğ»Ğ¸, ÑĞºÑ– Ğ°Ğ³ĞµĞ½Ñ‚ ÑƒĞ¶Ğµ Ğ²Ñ–Ğ´Ğ²Ñ–Ğ´Ğ°Ğ²
        self.known_edges = set()         # Ğ’Ñ–Ğ´Ğ¾Ğ¼Ñ– Ğ´Ğ¾Ñ€Ğ¾Ğ³Ğ¸ (Ñ€ĞµĞ±Ñ€Ğ°) Ğ¼Ñ–Ğ¶ Ğ²ÑƒĞ·Ğ»Ğ°Ğ¼Ğ¸
        self.road_signs = {}             # Ğ”Ğ¾Ñ€Ğ¾Ğ¶Ğ½Ñ– Ğ·Ğ½Ğ°ĞºĞ¸ Ğ´Ğ»Ñ Ğ½Ğ°Ğ¿Ñ€ÑĞ¼ĞºÑ–Ğ² (Ğ·Ğ²Ñ–Ğ´ĞºĞ¸ â†’ ĞºÑƒĞ´Ğ¸)

    def tell_visited(self, node):
        """Ğ”Ğ¾Ğ´Ğ°Ğ²Ğ°Ğ½Ğ½Ñ Ğ²Ñ–Ğ´Ğ²Ñ–Ğ´Ğ°Ğ½Ğ¾Ğ³Ğ¾ Ğ²ÑƒĞ·Ğ»Ğ°."""
        self.visited_nodes.add(node)
        print(f"TELL: Ğ’ÑƒĞ·Ğ¾Ğ» {node} Ğ´Ğ¾Ğ´Ğ°Ğ½Ğ¾ Ğ´Ğ¾ Ğ²Ñ–Ğ´Ğ²Ñ–Ğ´Ğ°Ğ½Ğ¸Ñ….")

    def tell_edge(self, node1, node2):
        """Ğ”Ğ¾Ğ´Ğ°Ğ²Ğ°Ğ½Ğ½Ñ Ğ²Ñ–Ğ´Ğ¾Ğ¼Ğ¾Ğ³Ğ¾ Ñ€ĞµĞ±Ñ€Ğ° (Ğ´Ğ¾Ñ€Ğ¾Ğ³Ğ¸)."""
        self.known_edges.add((node1, node2))
        self.known_edges.add((node2, node1))  # ĞĞ±Ğ¸Ğ´Ğ²Ğ° Ğ½Ğ°Ğ¿Ñ€ÑĞ¼ĞºĞ¸
        print(f"TELL: Ğ”Ğ¾Ñ€Ğ¾Ğ³Ğ° {node1} <-> {node2} Ğ´Ğ¾Ğ´Ğ°Ğ½Ğ° Ğ² Ğ±Ğ°Ğ·Ñƒ Ğ·Ğ½Ğ°Ğ½ÑŒ.")

    def tell_sign(self, from_node, to_node, sign_info):
        """Ğ—Ğ±ĞµÑ€ĞµĞ¶ĞµĞ½Ğ½Ñ Ğ´Ğ¾Ñ€Ğ¾Ğ¶Ğ½ÑŒĞ¾Ğ³Ğ¾ Ğ·Ğ½Ğ°ĞºĞ° Ğ´Ğ»Ñ Ğ½Ğ°Ğ¿Ñ€ÑĞ¼ĞºÑƒ."""
        if from_node not in self.road_signs:
            self.road_signs[from_node] = {}
        self.road_signs[from_node][to_node] = sign_info
        print(f"TELL: Ğ”Ğ¾Ñ€Ğ¾Ğ¶Ğ½Ñ–Ğ¹ Ğ·Ğ½Ğ°Ğº {from_node} -> {to_node}: {sign_info} Ğ´Ğ¾Ğ´Ğ°Ğ½Ğ¾.")

    def ask_visited(self, node):
        """ĞŸĞµÑ€ĞµĞ²Ñ–Ñ€ĞºĞ°, Ñ‡Ğ¸ Ğ²ÑƒĞ·Ğ¾Ğ» ÑƒĞ¶Ğµ Ğ²Ñ–Ğ´Ğ²Ñ–Ğ´Ğ°Ğ½Ğ¾."""
        return node in self.visited_nodes

    def ask_possible_moves(self, node):
        """ĞÑ‚Ñ€Ğ¸Ğ¼Ğ°Ğ½Ğ½Ñ Ğ´Ğ¾ÑÑ‚ÑƒĞ¿Ğ½Ğ¸Ñ… Ğ½Ğ°Ğ¿Ñ€ÑĞ¼ĞºÑ–Ğ² Ñ€ÑƒÑ…Ñƒ Ğ· Ğ²ÑƒĞ·Ğ»Ğ°."""
        return self.road_signs.get(node, {})

    def ask_known_edges(self):
        """ĞÑ‚Ñ€Ğ¸Ğ¼Ğ°Ğ½Ğ½Ñ Ğ²ÑÑ–Ñ… Ğ²Ñ–Ğ´Ğ¾Ğ¼Ğ¸Ñ… Ğ´Ğ¾Ñ€Ñ–Ğ³."""
        return self.known_edges


# Ğ†Ğ½Ñ–Ñ†Ñ–Ğ°Ğ»Ñ–Ğ·Ğ°Ñ†Ñ–Ñ Ğ±Ğ°Ğ·Ğ¸ Ğ·Ğ½Ğ°Ğ½ÑŒ
kb = KnowledgeBase()
print("\n--- Ğ‘Ğ°Ğ·Ğ° Ğ·Ğ½Ğ°Ğ½ÑŒ Ğ°Ğ³ĞµĞ½Ñ‚Ğ° ---")
print(f"Ğ’Ñ–Ğ´Ğ²Ñ–Ğ´Ğ°Ğ½Ñ– Ğ²ÑƒĞ·Ğ»Ğ¸: {kb.visited_nodes}")
print(f"Ğ’Ñ–Ğ´Ğ¾Ğ¼Ñ– Ğ´Ğ¾Ñ€Ğ¾Ğ³Ğ¸: {kb.known_edges}")
print(f"Ğ”Ğ¾Ñ€Ğ¾Ğ¶Ğ½Ñ– Ğ·Ğ½Ğ°ĞºĞ¸: {kb.road_signs}")
print("--------------------------\n")

# ĞŸĞ¾Ğ²Ñ‚Ğ¾Ñ€Ğ½Ğµ Ğ·Ğ°Ğ²Ğ°Ğ½Ñ‚Ğ°Ğ¶ĞµĞ½Ğ½Ñ Ğ¼Ğ¾Ğ´ĞµĞ»Ñ– (ÑĞºÑ‰Ğ¾ Ğ¿Ğ¾Ñ‚Ñ€Ñ–Ğ±Ğ½Ğ¾ Ğ² ĞºĞ¾Ğ½Ñ‚ĞµĞºÑÑ‚Ñ– ĞºĞ¾Ğ´Ñƒ)
model = load_model('speed_limit_model.keras')
print("ĞœĞ¾Ğ´ĞµĞ»ÑŒ Ğ¾Ğ±Ğ¼ĞµĞ¶ĞµĞ½Ğ½Ñ ÑˆĞ²Ğ¸Ğ´ĞºĞ¾ÑÑ‚Ñ– ÑƒÑĞ¿Ñ–ÑˆĞ½Ğ¾ Ğ·Ğ°Ğ²Ğ°Ğ½Ñ‚Ğ°Ğ¶ĞµĞ½Ğ°.")





    
# ğŸ‘ğŸ‘ğŸ‘ğŸ‘ Ğ’Ñ–Ğ·ÑƒĞ°Ğ»Ñ–Ğ·Ğ°Ñ†Ñ–Ñ Ñ€ÑƒÑ…Ñƒ ğŸ‘ğŸ‘ğŸ‘ğŸ‘


def draw_graph(G, agent_pos, visited, path, seen, speed_limits, start, end, previous_pos=None, current_sign_path=None, edge_speeds=None):
    """Ğ’Ñ–Ğ·ÑƒĞ°Ğ»Ñ–Ğ·Ğ°Ñ†Ñ–Ñ Ğ³Ñ€Ğ°Ñ„Ğ° Ğ· ĞºĞ¾Ğ»ÑŒĞ¾Ñ€Ğ¾Ğ²Ğ¸Ğ¼ Ğ²Ğ¸Ğ´Ñ–Ğ»ĞµĞ½Ğ½ÑĞ¼ Ğ²ÑƒĞ·Ğ»Ñ–Ğ² Ñ– Ğ²Ñ–Ğ´Ğ¾Ğ±Ñ€Ğ°Ğ¶ĞµĞ½Ğ½ÑĞ¼ Ğ·Ğ½Ğ°ĞºĞ° ÑˆĞ²Ğ¸Ğ´ĞºĞ¾ÑÑ‚Ñ–."""
    global last_sign_image  # Ğ—Ğ±ĞµÑ€ĞµĞ¶ĞµĞ½Ğ½Ñ Ğ¾ÑÑ‚Ğ°Ğ½Ğ½ÑŒĞ¾Ğ³Ğ¾ Ğ¿Ğ¾ĞºĞ°Ğ·Ğ°Ğ½Ğ¾Ğ³Ğ¾ Ğ·Ğ¾Ğ±Ñ€Ğ°Ğ¶ĞµĞ½Ğ½Ñ

    # ĞŸĞ¾Ğ±ÑƒĞ´Ğ¾Ğ²Ğ° Ğ¿Ğ¾Ğ·Ğ¸Ñ†Ñ–Ğ¹ Ğ²ÑƒĞ·Ğ»Ñ–Ğ² Ğ´Ğ»Ñ Ğ³Ñ€Ğ°Ñ„Ñ–ĞºĞ°
    pos = {(x, y): (x, -y) for x, y in G.nodes()}

    node_colors = []
    for node in G.nodes():
        if node == agent_pos:
            node_colors.append("red")  # ĞŸĞ¾Ñ‚Ğ¾Ñ‡Ğ½Ğ° Ğ¿Ğ¾Ğ·Ğ¸Ñ†Ñ–Ñ Ğ°Ğ³ĞµĞ½Ñ‚Ğ°
        elif node == end:
            node_colors.append("green")  # ĞšÑ–Ğ½Ñ†ĞµĞ²Ğ° Ñ‚Ğ¾Ñ‡ĞºĞ°
        elif node in path:
            node_colors.append("lightgreen")  # ĞŸÑ€Ğ¾Ğ¹Ğ´ĞµĞ½Ğ¸Ğ¹ ÑˆĞ»ÑÑ…
        elif node in seen:
            node_colors.append("yellow")  # ĞŸĞµÑ€ĞµĞ³Ğ»ÑĞ½ÑƒÑ‚Ñ–, Ğ°Ğ»Ğµ Ğ½Ğµ Ğ²Ñ–Ğ´Ğ²Ñ–Ğ´Ğ°Ğ½Ñ–
        elif node not in visited:
            node_colors.append("blue")  # ĞĞµ Ğ²Ñ–Ğ´Ğ²Ñ–Ğ´Ğ°Ğ½Ñ– Ñ– Ğ½Ğµ Ğ¿ĞµÑ€ĞµĞ³Ğ»ÑĞ½ÑƒÑ‚Ñ–
        else:
            node_colors.append("white")  # Ğ’Ñ–Ğ´Ğ²Ñ–Ğ´Ğ°Ğ½Ñ–

    plt.figure(figsize=(12, 6))

    # Ğ’Ñ–Ğ·ÑƒĞ°Ğ»Ñ–Ğ·Ğ°Ñ†Ñ–Ñ Ğ³Ñ€Ğ°Ñ„Ğ°
    plt.subplot(1, 2, 1)
    nx.draw(G, pos, with_labels=True, node_size=400, node_color=node_colors, edge_color="black")

    # Ğ”Ğ¾Ğ´Ğ°Ğ²Ğ°Ğ½Ğ½Ñ Ñ‚ĞµĞºÑÑ‚Ğ¾Ğ²Ğ¸Ñ… Ğ¿Ñ–Ğ´Ğ¿Ğ¸ÑÑ–Ğ² ÑˆĞ²Ğ¸Ğ´ĞºĞ¾ÑÑ‚Ñ– Ğ½Ğ° Ñ€ĞµĞ±Ñ€Ğ°Ñ…
    if edge_speeds:
        edge_labels = {(min(edge), max(edge)): f"{speed} ĞºĞ¼/Ğ³Ğ¾Ğ´" for edge, speed in edge_speeds.items()}
        nx.draw_networkx_edge_labels(G, pos, edge_labels=edge_labels, font_color='red')

    plt.title(f"ĞĞ³ĞµĞ½Ñ‚ Ğ·Ğ½Ğ°Ñ…Ğ¾Ğ´Ğ¸Ñ‚ÑŒÑÑ Ñƒ {agent_pos}")

    # ĞŸÑ–Ğ´Ğ³Ğ¾Ñ‚Ğ¾Ğ²ĞºĞ° Ğ´Ğ¾ Ğ²Ñ–Ğ´Ğ¾Ğ±Ñ€Ğ°Ğ¶ĞµĞ½Ğ½Ñ Ğ·Ğ¾Ğ±Ñ€Ğ°Ğ¶ĞµĞ½Ğ½Ñ Ğ·Ğ½Ğ°ĞºĞ°
    sign_image_path = current_sign_path

    # Ğ¡Ğ¿Ñ€Ğ¾Ğ±Ğ° Ğ¿Ğ¾ĞºĞ°Ğ·Ğ°Ñ‚Ğ¸ Ğ¿Ğ¾Ñ‚Ğ¾Ñ‡Ğ½Ğ¸Ğ¹ Ğ·Ğ½Ğ°Ğº
    if current_sign_path and os.path.exists(current_sign_path):
        last_sign_image = current_sign_path
        print(f"Ğ—Ğ°Ğ²Ğ°Ğ½Ñ‚Ğ°Ğ¶ĞµĞ½Ğ½Ñ Ğ·Ğ¾Ğ±Ñ€Ğ°Ğ¶ĞµĞ½Ğ½Ñ: {current_sign_path}")
        try:
            sign_image = cv2.imread(current_sign_path, cv2.IMREAD_GRAYSCALE)
            if sign_image is not None and sign_image.size > 0:
                last_sign_image = current_sign_path  # Ğ—Ğ±ĞµÑ€ĞµĞ¶ĞµĞ½Ğ½Ñ Ğ¾ÑÑ‚Ğ°Ğ½Ğ½ÑŒĞ¾Ğ³Ğ¾ Ğ²Ğ¸ĞºĞ¾Ñ€Ğ¸ÑÑ‚Ğ°Ğ½Ğ¾Ğ³Ğ¾ Ğ·Ğ½Ğ°ĞºĞ°
                sign_image = cv2.bitwise_not(sign_image)  # Ğ†Ğ½Ğ²ĞµÑ€ÑÑ–Ñ Ğ´Ğ»Ñ ĞºÑ€Ğ°Ñ‰Ğ¾Ğ³Ğ¾ ĞºĞ¾Ğ½Ñ‚Ñ€Ğ°ÑÑ‚Ñƒ
                sign_image = cv2.resize(sign_image, (100, 100), interpolation=cv2.INTER_AREA)

                # ĞŸĞ¾ĞºĞ°Ğ· Ğ·Ğ¾Ğ±Ñ€Ğ°Ğ¶ĞµĞ½Ğ½Ñ
                plt.subplot(1, 2, 2)
                plt.imshow(sign_image, cmap='gray', vmin=0, vmax=255)
                plt.axis('off')
                plt.title(f"Ğ—Ğ½Ğ°Ğº ÑˆĞ²Ğ¸Ğ´ĞºĞ¾ÑÑ‚Ñ– Ğ´Ğ»Ñ {agent_pos}")
            else:
                print(f"âš  Ğ—Ğ¾Ğ±Ñ€Ğ°Ğ¶ĞµĞ½Ğ½Ñ Ğ¿Ğ¾Ñ€Ğ¾Ğ¶Ğ½Ñ” Ğ°Ğ±Ğ¾ Ğ½Ğµ Ğ·Ğ½Ğ°Ğ¹Ğ´ĞµĞ½Ğ¾: {current_sign_path}")
        except Exception as e:
            print(f"âŒ ĞŸĞ¾Ğ¼Ğ¸Ğ»ĞºĞ° Ğ¾Ğ±Ñ€Ğ¾Ğ±ĞºĞ¸ Ğ·Ğ¾Ğ±Ñ€Ğ°Ğ¶ĞµĞ½Ğ½Ñ {current_sign_path}: {e}")

    # Ğ¯ĞºÑ‰Ğ¾ Ğ¿Ğ¾Ñ‚Ğ¾Ñ‡Ğ½Ğ¾Ğ³Ğ¾ Ğ·Ğ½Ğ°ĞºĞ° Ğ½ĞµĞ¼Ğ°Ñ” â€” Ğ¿Ğ¾ĞºĞ°Ğ· Ğ¾ÑÑ‚Ğ°Ğ½Ğ½ÑŒĞ¾Ğ³Ğ¾
    elif last_sign_image and os.path.exists(last_sign_image):
        print(f"Ğ’Ğ¸ĞºĞ¾Ñ€Ğ¸ÑÑ‚Ğ°Ğ½Ğ½Ñ Ğ¿Ğ¾Ğ¿ĞµÑ€ĞµĞ´Ğ½ÑŒĞ¾Ğ³Ğ¾ Ğ·Ğ½Ğ°ĞºĞ°: {last_sign_image}")
        try:
            sign_image = cv2.imread(last_sign_image, cv2.IMREAD_GRAYSCALE)
            if sign_image is not None and sign_image.size > 0:
                sign_image = cv2.bitwise_not(sign_image)
                sign_image = cv2.resize(sign_image, (100, 100), interpolation=cv2.INTER_AREA)

                plt.subplot(1, 2, 2)
                plt.imshow(sign_image, cmap='gray', vmin=0, vmax=255)
                plt.axis('off')
                plt.title("ĞÑÑ‚Ğ°Ğ½Ğ½Ñ–Ğ¹ Ğ²Ğ¸ĞºĞ¾Ñ€Ğ¸ÑÑ‚Ğ°Ğ½Ğ¸Ğ¹ Ğ·Ğ½Ğ°Ğº")
            else:
                print(f"âš  Ğ—Ğ¾Ğ±Ñ€Ğ°Ğ¶ĞµĞ½Ğ½Ñ Ğ¿Ğ¾Ñ€Ğ¾Ğ¶Ğ½Ñ” Ğ°Ğ±Ğ¾ Ğ½Ğµ Ğ·Ğ½Ğ°Ğ¹Ğ´ĞµĞ½Ğ¾: {last_sign_image}")
        except Exception as e:
            print(f"âŒ ĞŸĞ¾Ğ¼Ğ¸Ğ»ĞºĞ° Ğ¾Ğ±Ñ€Ğ¾Ğ±ĞºĞ¸ Ğ·Ğ¾Ğ±Ñ€Ğ°Ğ¶ĞµĞ½Ğ½Ñ {last_sign_image}: {e}")
    else:
        print("âš  ĞĞµĞ¼Ğ°Ñ” Ğ´Ğ¾ÑÑ‚ÑƒĞ¿Ğ½Ğ¾Ğ³Ğ¾ Ğ·Ğ¾Ğ±Ñ€Ğ°Ğ¶ĞµĞ½Ğ½Ñ Ğ´Ğ»Ñ Ğ²Ñ–Ğ´Ğ¾Ğ±Ñ€Ğ°Ğ¶ĞµĞ½Ğ½Ñ.")

    plt.show()






# ğŸ•¹ğŸ•¹ğŸ•¹ğŸ•¹ Ğ›Ğ¾Ğ³Ñ–ĞºĞ° Ñ€ÑƒÑ…Ñƒ Ğ°Ğ³ĞµĞ½Ñ‚Ğ° ğŸ•¹ğŸ•¹ğŸ•¹ğŸ•¹



def move_agent_with_knowledge(G, kb, start, end, max_steps=50):
    """Ğ¤ÑƒĞ½ĞºÑ†Ñ–Ñ Ğ´Ğ»Ñ Ñ€ÑƒÑ…Ñƒ Ğ°Ğ³ĞµĞ½Ñ‚Ğ° Ğ· Ğ²Ğ¸ĞºĞ¾Ñ€Ğ¸ÑÑ‚Ğ°Ğ½Ğ½ÑĞ¼ Ğ±Ğ°Ğ·Ğ¸ Ğ·Ğ½Ğ°Ğ½ÑŒ Ñ‚Ğ° ĞºĞ¾Ğ½Ñ‚Ñ€Ğ¾Ğ»Ñ ÑˆĞ²Ğ¸Ğ´ĞºĞ¾ÑÑ‚Ñ–."""

    # Ğ†Ğ½Ñ–Ñ†Ñ–Ğ°Ğ»Ñ–Ğ·Ğ°Ñ†Ñ–Ñ Ğ·Ğ¼Ñ–Ğ½Ğ½Ğ¸Ñ…
    visited = set()
    path = []
    seen = deque()  # Ğ”Ğ²Ğ¾ÑÑ‚Ğ¾Ñ€Ğ¾Ğ½Ğ½Ñ Ñ‡ĞµÑ€Ğ³Ğ° Ğ´Ğ»Ñ Ğ·Ğ±ĞµÑ€ĞµĞ¶ĞµĞ½Ğ½Ñ Ğ±Ğ°Ñ‡ĞµĞ½Ğ¸Ñ… Ğ²ÑƒĞ·Ğ»Ñ–Ğ²
    avoid_nodes = set()  # ĞĞ°Ğ±Ñ–Ñ€ Ğ²ÑƒĞ·Ğ»Ñ–Ğ², ÑĞºĞ¸Ñ… Ğ¿Ğ¾Ñ‚Ñ€Ñ–Ğ±Ğ½Ğ¾ ÑƒĞ½Ğ¸ĞºĞ°Ñ‚Ğ¸
    current_pos = start
    previous_pos = None
    direction = "right"
    step_count = 0
    in_dead_end = False
    visited_nodes_counter = {}
    skip_speed_check = False

    # Ğ¡Ğ»Ğ¾Ğ²Ğ½Ğ¸Ğº Ğ´Ğ»Ñ Ğ·Ğ±ĞµÑ€ĞµĞ¶ĞµĞ½Ğ½Ñ Ğ¾Ğ±Ğ¼ĞµĞ¶ĞµĞ½ÑŒ ÑˆĞ²Ğ¸Ğ´ĞºĞ¾ÑÑ‚Ñ– Ğ·Ğ° Ğ´Ğ¾Ñ€Ğ¾Ğ³Ğ°Ğ¼Ğ¸ (Ğ·Ğ²â€™ÑĞ·ĞºĞ°Ğ¼Ğ¸)
    speed_limits = {}

    def get_speed_limit(node1, node2):
        """Ğ¤ÑƒĞ½ĞºÑ†Ñ–Ñ Ğ´Ğ»Ñ Ğ¾Ñ‚Ñ€Ğ¸Ğ¼Ğ°Ğ½Ğ½Ñ ÑˆĞ²Ğ¸Ğ´ĞºĞ¾ÑÑ‚Ñ– Ğ´Ğ»Ñ Ğ¿ĞµĞ²Ğ½Ğ¾Ñ— Ğ´Ğ¾Ñ€Ğ¾Ğ³Ğ¸."""
        edge_key = (previous_pos, current_pos)
        if edge_key not in speed_limits:
            speed_limits[edge_key] = random.choice(range(20, 100, 10))  # Ğ’Ğ¸Ğ¿Ğ°Ğ´ĞºĞ¾Ğ²Ğ° ÑˆĞ²Ğ¸Ğ´ĞºÑ–ÑÑ‚ÑŒ
        return speed_limits[edge_key]


     # --- ĞÑĞ½Ğ¾Ğ²Ğ½Ğ¸Ğ¹ Ñ†Ğ¸ĞºĞ» Ñ€ÑƒÑ…Ñƒ ---
    while current_pos != end and step_count < max_steps:
        # ĞĞ½Ğ¾Ğ²Ğ»ĞµĞ½Ğ½Ñ ÑÑ‚Ğ°Ğ½Ñƒ Ğ°Ğ³ĞµĞ½Ñ‚Ğ°
        visited.add(current_pos)
        path.append(current_pos)
        kb.tell_visited(current_pos)  # Ğ”Ğ¾Ğ´Ğ°Ğ²Ğ°Ğ½Ğ½Ñ Ğ¿Ğ¾Ñ‚Ğ¾Ñ‡Ğ½Ğ¾Ğ³Ğ¾ Ğ²ÑƒĞ·Ğ»Ğ° Ğ´Ğ¾ Ğ±Ğ°Ğ·Ğ¸ Ğ·Ğ½Ğ°Ğ½ÑŒ
        print(f"ğŸ“¥ Ğ‘Ğ°Ğ·Ğ° Ğ·Ğ½Ğ°Ğ½ÑŒ Ğ¾Ğ½Ğ¾Ğ²Ğ»ĞµĞ½Ğ°: Ğ´Ğ¾Ğ´Ğ°Ğ½Ğ¾ Ñ–Ğ½Ñ„Ğ¾Ñ€Ğ¼Ğ°Ñ†Ñ–Ñ Ğ¿Ñ€Ğ¾ Ğ²ÑƒĞ·Ğ¾Ğ» {current_pos}")
        step_count += 1

        # ĞĞ½Ğ°Ğ»Ñ–Ğ· ÑÑƒÑÑ–Ğ´Ğ½Ñ–Ñ… Ğ²ÑƒĞ·Ğ»Ñ–Ğ²
        neighbors = list(G.neighbors(current_pos))
        print(f"ğŸ” Ğ¡ĞºĞ°Ğ½ÑƒĞ²Ğ°Ğ½Ğ½Ñ ÑÑƒÑÑ–Ğ´Ğ½Ñ–Ñ… Ğ¿ĞµÑ€ĞµÑ…Ñ€ĞµÑÑ‚ÑŒ Ğ´Ğ»Ñ Ğ²ÑƒĞ·Ğ»Ğ° {current_pos}: {neighbors}")

        for neighbor in neighbors:
            kb.tell_edge(current_pos, neighbor)
            direction_sign = get_direction(current_pos, neighbor)
            kb.tell_sign(current_pos, neighbor, direction_sign)
            print(f"ğŸš¸ Ğ”Ğ¾Ñ€Ğ¾Ğ¶Ğ½Ñ–Ğ¹ Ğ·Ğ½Ğ°Ğº: Ğ· {current_pos} Ñƒ {neighbor} -> {direction_sign}")

            # ĞŸĞµÑ€ĞµĞ²Ñ–Ñ€ĞºĞ° Ğ½Ğ° Ñ‚ÑƒĞ¿Ğ¸ĞºĞ¸: ÑĞºÑ‰Ğ¾ Ñ” Ğ»Ğ¸ÑˆĞµ Ğ¾Ğ´Ğ¸Ğ½ ÑˆĞ»ÑÑ… Ğ½Ğ°Ğ·Ğ°Ğ´, ÑƒĞ½Ğ¸ĞºĞ°Ñ”Ğ¼Ğ¾
            if len(list(G.neighbors(neighbor))) == 1 and neighbor != end:
                avoid_nodes.add(neighbor)
                print(f"âš  Ğ’Ğ¸ÑĞ²Ğ»ĞµĞ½Ğ¾ Ñ‚ÑƒĞ¿Ğ¸ĞºĞ¾Ğ²Ğ¸Ğ¹ Ğ²ÑƒĞ·Ğ¾Ğ»: {neighbor}, Ğ²Ñ–Ğ½ Ğ±ÑƒĞ´Ğµ ÑƒĞ½Ğ¸ĞºĞ°Ñ‚Ğ¸ÑÑŒ.")


            # Ğ”Ğ¾Ğ´Ğ°Ğ²Ğ°Ğ½Ğ½Ñ Ğ²ÑƒĞ·Ğ»Ğ° Ğ´Ğ¾ Ğ±Ğ°Ñ‡ĞµĞ½Ğ¸Ñ…
            if neighbor not in visited and neighbor not in seen:
                seen.append(neighbor)
                print(f"ğŸ‘€ Ğ”Ğ¾Ğ´Ğ°Ğ½Ğ¾ Ğ´Ğ¾ Ğ±Ğ°Ñ‡ĞµĞ½Ğ¸Ñ… Ğ²ÑƒĞ·Ğ»Ñ–Ğ²: {neighbor}")

        # Ğ’Ğ¸Ğ´Ğ°Ğ»ĞµĞ½Ğ½Ñ Ğ¿Ğ¾Ñ‚Ğ¾Ñ‡Ğ½Ğ¾Ñ— Ğ¿Ğ¾Ğ·Ğ¸Ñ†Ñ–Ñ— Ğ· Ñ‡ĞµÑ€Ğ³Ğ¸ Ğ±Ğ°Ñ‡ĞµĞ½Ğ¸Ñ…
        if current_pos in seen:
            seen.remove(current_pos)

        from IPython.display import display, Markdown
        
        # --- Ğ’Ñ–Ğ·ÑƒĞ°Ğ»ÑŒĞ½Ğ¸Ğ¹ Ğ¾Ğ¿Ğ¸Ñ ĞºÑ€Ğ¾ĞºÑƒ ---
        step_info = f"ğŸš— ĞšÑ€Ğ¾Ğº {step_count}: ĞĞ³ĞµĞ½Ñ‚ Ñ€ÑƒÑ…Ğ°Ñ”Ñ‚ÑŒÑÑ Ğ´Ğ¾ {current_pos}"
        display(Markdown(f"<span style='font-size:20px; font-weight:bold'>{step_info}</span>"))

        print(f"ğŸ§­ ĞŸĞ¾Ñ‚Ğ¾Ñ‡Ğ½Ğ¸Ğ¹ Ğ½Ğ°Ğ¿Ñ€ÑĞ¼Ğ¾Ğº: {direction}")

        # --- 
        if previous_pos is not None:
            edge_key = (previous_pos, current_pos)
            current_sign_path = road_signs.get(edge_key, f'signs/speed_sign_0.png')
            
            #draw_graph(G, current_pos, visited, path, seen, speed_limits, start, end, previous_pos, current_sign_path)


            

        # ĞÑ‚Ñ€Ğ¸Ğ¼Ğ°Ğ½Ğ½Ñ Ğ¾Ğ±Ğ¼ĞµĞ¶ĞµĞ½Ğ½Ñ ÑˆĞ²Ğ¸Ğ´ĞºĞ¾ÑÑ‚Ñ– Ğ´Ğ»Ñ Ğ´Ğ¾Ñ€Ğ¾Ğ³Ğ¸
        if previous_pos is not None and not skip_speed_check:
            edge_key = (previous_pos, current_pos)
            sign_image_path = road_signs.get(edge_key)
            if sign_image_path is None:
                print(f"âš  ĞĞµĞ¼Ğ°Ñ” Ğ·Ğ¾Ğ±Ñ€Ğ°Ğ¶ĞµĞ½Ğ½Ñ Ğ´Ğ»Ñ Ñ€ĞµĞ±Ñ€Ğ° {edge_key}, Ğ²Ğ¸ĞºĞ¾Ñ€Ğ¸ÑÑ‚Ğ¾Ğ²ÑƒÑ Ğ·Ğ°Ğ³Ğ»ÑƒÑˆĞºÑƒ.")
                sign_image_path = 'signs/speed_sign_0.png'
        
            speed_limit = recognize_speed_limit(sign_image_path, model)
            speed_limits[edge_key] = speed_limit
            print(f"ğŸš¦ ĞĞ±Ğ¼ĞµĞ¶ĞµĞ½Ğ½Ñ ÑˆĞ²Ğ¸Ğ´ĞºĞ¾ÑÑ‚Ñ– Ğ´Ğ»Ñ Ğ´Ğ¾Ñ€Ğ¾Ğ³Ğ¸ {previous_pos} -> {current_pos}: {speed_limit} ĞºĞ¼/Ğ³Ğ¾Ğ´")
        else:
            speed_limit = 50
            sign_image_path = 'signs/speed_sign_0.png'  # ğŸ›  Ğ”ĞĞ‘ĞĞ’Ğ›Ğ•ĞĞ: Ğ´ĞµÑ„Ğ¾Ğ»Ñ‚Ğ½Ğµ Ğ·Ğ¾Ğ±Ñ€Ğ°Ğ¶ĞµĞ½Ğ½Ñ

  
        #sign_image_path = None 
    
        # Ğ’Ñ–Ğ´Ğ¾Ğ±Ñ€Ğ°Ğ¶ĞµĞ½Ğ½Ñ Ğ³Ñ€Ğ°Ñ„Ğ° Ğ· Ğ´Ğ¾Ğ´Ğ°Ğ½Ğ¸Ğ¼ Ğ·Ğ¾Ğ±Ñ€Ğ°Ğ¶ĞµĞ½Ğ½ÑĞ¼ Ğ·Ğ½Ğ°ĞºĞ°
        pass            
        current_sign_path = sign_image_path  


            

        # Ğ’Ğ¸ÑĞ²Ğ»ĞµĞ½Ğ½Ñ Ñ‚ÑƒĞ¿Ğ¸ĞºÑ–Ğ² Ñ– Ñ†Ğ¸ĞºĞ»Ñ–Ğ²
        visited_nodes_counter[current_pos] = visited_nodes_counter.get(current_pos, 0) + 1
        if visited_nodes_counter[current_pos] > 2:
            in_dead_end = True
            print(f"ğŸ”„ Ğ¦Ğ˜ĞšĞ› Ğ’Ğ˜Ğ¯Ğ’Ğ›Ğ•ĞĞ! Ğ’Ğ¥Ğ†Ğ” Ğ£ Ğ¡Ğ¢ĞĞ Ğ¢Ğ£ĞŸĞ˜ĞšĞ Ğ½Ğ° Ğ²ÑƒĞ·Ğ»Ñ– {current_pos}")

        # Ğ£Ğ½Ğ¸ĞºĞ½ĞµĞ½Ğ½Ñ Ñ‚ÑƒĞ¿Ğ¸ĞºÑ–Ğ² Ñ‚Ğ° Ñ€ÑƒÑ… Ğ´Ğ¾ Ğ±Ğ°Ñ‡ĞµĞ½Ğ¸Ñ… Ğ²ÑƒĞ·Ğ»Ñ–Ğ²
        if (len(neighbors) == 1 and previous_pos in neighbors) or in_dead_end:
            in_dead_end = True
            if seen:
                target_seen = next((node for node in reversed(seen) if node not in avoid_nodes), None)
                print(f"â†© Ğ¢Ğ£ĞŸĞ˜Ğš! ĞĞ³ĞµĞ½Ñ‚ Ğ¿Ñ€ÑĞ¼ÑƒÑ” Ğ¿Ğ¾ĞºÑ€Ğ¾ĞºĞ¾Ğ²Ğ¾ Ğ´Ğ¾ Ğ¶Ğ¾Ğ²Ñ‚Ğ¾Ñ— Ñ‚Ğ¾Ñ‡ĞºĞ¸ {target_seen}")

                if target_seen:
                    try:
                        shortest_path = nx.shortest_path(G, current_pos, target_seen)
                        next_pos = shortest_path[1]

                        # ğŸ§  ĞĞĞĞ’Ğ›Ğ•ĞĞĞ¯ Ğ¡Ğ¢ĞĞĞ£
                        new_direction = get_direction(current_pos, next_pos)
                        direction = new_direction
                        previous_pos = current_pos

                        
                        current_pos = next_pos

                        # ğŸ§© Ğ’Ğ†Ğ—Ğ£ĞĞ›Ğ†Ğ—ĞĞ¦Ğ†Ğ¯ ĞŸĞ•Ğ Ğ•Ğ” ĞŸĞ ĞĞ”ĞĞ’Ğ–Ğ•ĞĞĞ¯Ğœ
                        # Ğ’Ğ¸Ğ·Ğ½Ğ°Ñ‡ĞµĞ½Ğ½Ñ Ğ·Ğ½Ğ°ĞºÑƒ
                        edge_key = (previous_pos, current_pos)
                        #sign_image_path = road_signs.get(edge_key, 'signs/speed_sign_0.png')
                        current_sign_path = sign_image_path
                        
                        # Ğ”Ğ¾Ğ´Ğ°Ğ²Ğ°Ğ½Ğ½Ñ Ğ´Ğ¾ Ğ³Ñ€Ğ°Ñ„Ğ° (ÑĞºÑ‰Ğ¾ Ğ¿Ğ¾Ñ‚Ñ€Ñ–Ğ±Ğ½Ğ¾ â€” Ğ¾Ğ½Ğ¾Ğ²Ğ¸Ñ‚Ğ¸ ÑˆĞ²Ğ¸Ğ´ĞºÑ–ÑÑ‚ÑŒ)
                        """if edge_key not in speed_limits:
                            speed_limit = recognize_speed_limit(sign_image_path, model)
                            speed_limits[edge_key] = speed_limit"""
                        
                        draw_graph(G, current_pos, visited, path, seen, speed_limits, start, end, previous_pos, current_sign_path)
                        
                        continue 
                    except nx.NetworkXNoPath:
                        print("âŒ ĞĞµĞ¼Ğ°Ñ” ÑˆĞ»ÑÑ…Ñƒ Ğ´Ğ¾ Ğ¶Ğ¾Ğ²Ñ‚Ğ¾Ñ— Ñ‚Ğ¾Ñ‡ĞºĞ¸, Ğ·Ğ°Ğ²ĞµÑ€ÑˆĞµĞ½Ğ½Ñ Ñ€ÑƒÑ…Ñƒ.")
                        break
                else:
                    print("âŒ ĞĞµĞ¼Ğ°Ñ” Ğ¶Ğ¾Ğ²Ñ‚Ğ¸Ñ… Ñ‚Ğ¾Ñ‡Ğ¾Ğº Ğ´Ğ»Ñ Ğ¿Ğ¾Ğ²ĞµÑ€Ğ½ĞµĞ½Ğ½Ñ, Ğ·Ğ°Ğ²ĞµÑ€ÑˆĞµĞ½Ğ½Ñ Ñ€ÑƒÑ…Ñƒ.")
                    break
        else:
            new_moves, old_moves = analyze_moves(G, current_pos, visited)
            possible_moves = [node for node in new_moves + old_moves if node not in avoid_nodes]

            # Ğ¯ĞºÑ‰Ğ¾ ĞºÑ–Ğ½Ñ†ĞµĞ²Ğ° Ñ‚Ğ¾Ñ‡ĞºĞ° Ñ” ÑĞµÑ€ĞµĞ´ Ğ±Ğ°Ñ‡ĞµĞ½Ğ¸Ñ…, Ñ€ÑƒÑ… Ğ¿Ğ¾ĞºÑ€Ğ¾ĞºĞ¾Ğ²Ğ¸Ğ¹
        if end in seen:
            print(f"ğŸ¯ ĞšÑ–Ğ½Ñ†ĞµĞ²Ğ° Ñ‚Ğ¾Ñ‡ĞºĞ° {end} Ğ±ÑƒĞ»Ğ° Ğ¿ĞµÑ€ĞµĞ³Ğ»ÑĞ½ÑƒÑ‚Ğ°, Ñ€ÑƒÑ…Ğ°Ñ”Ğ¼Ğ¾ÑÑ Ğ¿Ğ¾ĞºÑ€Ğ¾ĞºĞ¾Ğ²Ğ¾!")
            shortest_path = nx.shortest_path(G, current_pos, end)
            next_pos = shortest_path[1]
        else:
            next_pos = min(possible_moves, key=lambda pos: abs(pos[0] - end[0]) + abs(pos[1] - end[1]))





        # ĞĞ½Ğ¾Ğ²Ğ»ĞµĞ½Ğ½Ñ Ğ½Ğ°Ğ¿Ñ€ÑĞ¼ĞºÑƒ Ñ€ÑƒÑ…Ñƒ
        new_direction = get_direction(current_pos, next_pos)
        if new_direction != direction:
            if (direction, new_direction) in [("up", "right"), ("right", "down"), ("down", "left"), ("left", "up")]:
                print("â¡ ĞŸĞ¾Ğ²Ğ¾Ñ€Ğ¾Ñ‚ Ğ½Ğ°Ğ¿Ñ€Ğ°Ğ²Ğ¾")
            elif (direction, new_direction) in [("up", "left"), ("left", "down"), ("down", "right"), ("right", "up")]:
                print("â¬… ĞŸĞ¾Ğ²Ğ¾Ñ€Ğ¾Ñ‚ Ğ½Ğ°Ğ»Ñ–Ğ²Ğ¾")
            else:
            
                print("ğŸ”„ Ğ Ğ¾Ğ·Ğ²Ğ¾Ñ€Ğ¾Ñ‚ Ğ½Ğ° 180 Ğ³Ñ€Ğ°Ğ´ÑƒÑÑ–Ğ²")



        # Ğ¡Ğ¿Ğ¾Ñ‡Ğ°Ñ‚ĞºÑƒ Ğ²Ğ¸ĞºĞ»Ğ¸Ğº Ğ²Ñ–Ğ´Ğ¾Ğ±Ñ€Ğ°Ğ¶ĞµĞ½Ğ½Ñ Ğ³Ñ€Ğ°Ñ„Ğ° Ğ· Ğ¿Ñ€Ğ°Ğ²Ğ¸Ğ»ÑŒĞ½Ğ¸Ğ¼ Ğ·Ğ½Ğ°Ñ‡ĞµĞ½Ğ½ÑĞ¼ Ğ·Ğ½Ğ°ĞºĞ°
        if previous_pos is not None and not skip_speed_check:
            edge_key = (previous_pos, current_pos)
            if edge_key in road_signs:
                sign_image_path = road_signs.get(edge_key)
            else:
                print(f"âš  ĞĞµĞ¼Ğ°Ñ” Ğ·Ğ¾Ğ±Ñ€Ğ°Ğ¶ĞµĞ½Ğ½Ñ Ğ´Ğ»Ñ Ñ€ĞµĞ±Ñ€Ğ° {edge_key}, Ğ²Ğ¸ĞºĞ¾Ñ€Ğ¸ÑÑ‚Ğ¾Ğ²ÑƒÑ Ğ·Ğ°Ğ³Ğ»ÑƒÑˆĞºÑƒ.")
                sign_image_path = 'signs/speed_sign_0.png'
        else:
            sign_image_path = f'signs/speed_sign_0.png'

        current_sign_path = sign_image_path
            
        draw_graph(G, current_pos, visited, path, seen, speed_limits, start, end, previous_pos, current_sign_path)


        

        direction = new_direction
        previous_pos = current_pos
        current_pos = next_pos
        skip_speed_check = False






    # ĞšÑ–Ğ½Ñ†ĞµĞ²Ğ° Ñ‚Ğ¾Ñ‡ĞºĞ°, ÑĞºÑ‰Ğ¾ Ğ´Ğ¾ÑÑĞ³Ğ»Ğ¸
    if current_pos == end:
        kb.tell_visited(current_pos)
        path.append(current_pos)
        visited.add(current_pos)
        print(f"ğŸ Ğ”Ğ¾ÑÑĞ³Ğ½ÑƒÑ‚Ğ¾ ĞºÑ–Ğ½Ñ†ĞµĞ²Ğ¾Ñ— Ñ‚Ğ¾Ñ‡ĞºĞ¸! Ğ”Ğ¾Ğ´Ğ°Ğ½Ğ° {current_pos} Ğ´Ğ¾ visited_nodes.")

        # Ğ¤Ñ–Ğ½Ğ°Ğ»ÑŒĞ½Ğ° Ğ²Ñ–Ğ·ÑƒĞ°Ğ»Ñ–Ğ·Ğ°Ñ†Ñ–Ñ
        edge_key = (previous_pos, current_pos)
        sign_image_path = road_signs.get(edge_key, 'signs/speed_sign_0.png')
        current_sign_path = sign_image_path
    
        draw_graph(G, current_pos, visited, path, seen, speed_limits, start=end, end=end, previous_pos=previous_pos, current_sign_path=current_sign_path)



    print("âœ… Ğ ÑƒÑ… Ğ·Ğ°Ğ²ĞµÑ€ÑˆĞµĞ½Ğ¾!")
    print(f"ğŸ“Š Ğ¤Ğ°ĞºÑ‚Ğ¸Ñ‡Ğ½Ğ° ĞºÑ–Ğ»ÑŒĞºÑ–ÑÑ‚ÑŒ ĞºÑ€Ğ¾ĞºÑ–Ğ² Ğ°Ğ³ĞµĞ½Ñ‚Ğ°: {step_count}")

    return visited, path, seen, speed_limits, step_count









    
# ğŸ§©ğŸ§©ğŸ§©ğŸ§© Ğ”Ğ¾Ğ¿Ğ¾Ğ¼Ñ–Ğ¶Ğ½Ñ– Ñ„ÑƒĞ½ĞºÑ†Ñ–Ñ— ğŸ§©ğŸ§©ğŸ§©ğŸ§©


def get_direction(from_pos, to_pos):
    """Ğ’Ğ¸Ğ·Ğ½Ğ°Ñ‡ĞµĞ½Ğ½Ñ Ğ½Ğ°Ğ¿Ñ€ÑĞ¼ĞºÑƒ Ñ€ÑƒÑ…Ñƒ Ğ°Ğ³ĞµĞ½Ñ‚Ğ°."""
    dx, dy = to_pos[0] - from_pos[0], to_pos[1] - from_pos[1]
    if dx == 1: return "right"
    if dx == -1: return "left"
    if dy == -1: return "up"
    if dy == 1: return "down"
    return None


def analyze_moves(G, current_pos, visited):
    """ĞĞ½Ğ°Ğ»Ñ–Ğ· Ğ¼Ğ¾Ğ¶Ğ»Ğ¸Ğ²Ğ¸Ñ… Ñ€ÑƒÑ…Ñ–Ğ² Ğ°Ğ³ĞµĞ½Ñ‚Ğ°."""
    neighbors = list(G.neighbors(current_pos))
    new_moves = [node for node in neighbors if node not in visited]
    old_moves = [node for node in neighbors if node in visited]
    return new_moves, old_moves


def print_knowledge_base(kb):
    """Ğ’Ğ¸Ğ²Ñ–Ğ´ Ğ¿Ğ¾Ñ‚Ğ¾Ñ‡Ğ½Ğ¾Ğ³Ğ¾ ÑÑ‚Ğ°Ğ½Ñƒ Ğ±Ğ°Ğ·Ğ¸ Ğ·Ğ½Ğ°Ğ½ÑŒ Ğ°Ğ³ĞµĞ½Ñ‚Ğ°."""
    print("\n--- Ğ‘Ğ°Ğ·Ğ° Ğ·Ğ½Ğ°Ğ½ÑŒ Ğ°Ğ³ĞµĞ½Ñ‚Ğ° ---")
    print(f"Ğ’Ñ–Ğ´Ğ²Ñ–Ğ´Ğ°Ğ½Ñ– Ğ²ÑƒĞ·Ğ»Ğ¸: {kb.visited_nodes}")
    print(f"Ğ’Ñ–Ğ´Ğ¾Ğ¼Ñ– Ğ´Ğ¾Ñ€Ğ¾Ğ³Ğ¸: {kb.known_edges}")
    print(f"Ğ”Ğ¾Ñ€Ğ¾Ğ¶Ğ½Ñ– Ğ·Ğ½Ğ°ĞºĞ¸:")
    for node, signs in kb.road_signs.items():
        print(f"  {node}: {signs}")
    print("--------------------------\n")
    

def _update_agent_state(current_pos, visited, path, kb):
    """ĞĞ½Ğ¾Ğ²Ğ»ĞµĞ½Ğ½Ñ ÑÑ‚Ğ°Ğ½Ñƒ Ğ°Ğ³ĞµĞ½Ñ‚Ğ° (Ğ´Ğ¾Ğ´Ğ°Ğ²Ğ°Ğ½Ğ½Ñ Ñƒ Ğ±Ğ°Ğ·Ñƒ Ğ·Ğ½Ğ°Ğ½ÑŒ)."""
    visited.add(current_pos)
    path.append(current_pos)
    kb.tell_visited(current_pos)


def _analyze_neighbors(G, current_pos, kb, seen, avoid_nodes, visited):
    """ĞĞ½Ğ°Ğ»Ñ–Ğ· ÑÑƒÑÑ–Ğ´Ğ½Ñ–Ñ… Ğ²ÑƒĞ·Ğ»Ñ–Ğ², Ğ´Ğ¾Ğ´Ğ°Ğ²Ğ°Ğ½Ğ½Ñ Ñ–Ğ½Ñ„Ğ¾Ñ€Ğ¼Ğ°Ñ†Ñ–Ñ— Ğ´Ğ¾ Ğ±Ğ°Ğ·Ğ¸ Ğ·Ğ½Ğ°Ğ½ÑŒ."""
    neighbors = list(G.neighbors(current_pos))
    for neighbor in neighbors:
        kb.tell_edge(current_pos, neighbor)
        direction_sign = get_direction(current_pos, neighbor)
        kb.tell_sign(current_pos, neighbor, direction_sign)

        # ĞŸĞµÑ€ĞµĞ²Ñ–Ñ€ĞºĞ° ÑÑƒÑÑ–Ğ´Ğ½Ñ–Ñ… Ğ²ÑƒĞ·Ğ»Ñ–Ğ² Ğ½Ğ° Ñ‚ÑƒĞ¿Ğ¸ĞºĞ¸ Ñ‚Ğ° ÑƒĞ½Ñ–ĞºĞ°Ğ»ÑŒĞ½Ñ–ÑÑ‚ÑŒ
        neighbor_neighbors = list(G.neighbors(neighbor))
        if len(neighbor_neighbors) == 1:
            avoid_nodes.add(neighbor)

        # Ğ”Ğ¾Ğ´Ğ°Ñ”Ğ²Ğ°Ğ½Ğ½Ñ Ğ²ÑƒĞ·Ğ»Ñ–Ğ² Ğ² ÑĞ¿Ğ¸ÑĞ¾Ğº Ğ±Ğ°Ñ‡ĞµĞ½Ğ¸Ñ…, ÑĞºÑ‰Ğ¾ Ğ²Ğ¾Ğ½Ğ¸ Ñ‰Ğµ Ğ½Ğµ Ğ²Ñ–Ğ´Ğ²Ñ–Ğ´Ğ°Ğ½Ñ–
        if neighbor not in visited and neighbor not in seen:
            seen.append(neighbor)

        # ĞĞ½Ğ°Ğ»Ñ–Ğ· ÑÑƒÑÑ–Ğ´Ğ½Ñ–Ñ… Ğ¿ĞµÑ€ĞµÑ…Ñ€ĞµÑÑ‚ÑŒ Ñ‚Ğ° Ğ·Ğ±ĞµÑ€ĞµĞ¶ĞµĞ½Ğ½Ñ Ñ—Ñ…Ğ½ÑŒĞ¾Ñ— ÑÑ‚Ñ€ÑƒĞºÑ‚ÑƒÑ€Ğ¸
        for neighbor_of_neighbor in G.neighbors(neighbor):
            if neighbor_of_neighbor != current_pos and neighbor_of_neighbor not in visited:
                kb.tell_sign(neighbor, neighbor_of_neighbor, get_direction(neighbor, neighbor_of_neighbor))
                seen.append(neighbor_of_neighbor)

    # Ğ’Ğ¸Ğ´Ğ°Ğ»ĞµĞ½Ğ½Ñ Ğ²Ñ–Ğ´Ğ²Ñ–Ğ´Ğ°Ğ½Ğ¸Ñ… Ğ²ÑƒĞ·Ğ»Ğ¸ Ğ·Ñ– ÑĞ¿Ğ¸ÑĞºÑƒ Ğ±Ğ°Ñ‡ĞµĞ½Ğ¸Ñ…
    seen = deque([node for node in seen if node not in visited])




def _print_status(step_count, current_pos, direction):
    """Ğ’Ğ¸Ğ²Ñ–Ğ´ Ğ¿Ğ¾Ñ‚Ğ¾Ñ‡Ğ½Ğ¾Ğ³Ğ¾ ÑÑ‚Ğ°Ñ‚ÑƒÑÑƒ Ğ°Ğ³ĞµĞ½Ñ‚Ğ°."""
    print(f"\nĞšÑ€Ğ¾Ğº {step_count}: ĞĞ³ĞµĞ½Ñ‚ Ñ€ÑƒÑ…Ğ°Ñ”Ñ‚ÑŒÑÑ Ğ´Ğ¾ {current_pos}")
    print(f"ĞŸĞ¾Ñ‚Ğ¾Ñ‡Ğ½Ğ¸Ğ¹ Ğ½Ğ°Ğ¿Ñ€ÑĞ¼Ğ¾Ğº: {direction}")



def _handle_decision(G, kb, current_pos, end, seen, avoid_nodes, visited, path, in_dead_end):
    """ĞĞ±Ñ€Ğ¾Ğ±ĞºĞ° Ğ´Ğ¾ÑÑĞ³Ğ½ĞµĞ½Ğ½Ñ ĞºÑ–Ğ½Ñ†ĞµĞ²Ğ¾Ñ— Ñ‚Ğ¾Ñ‡ĞºĞ¸ Ğ°Ğ±Ğ¾ Ñ€ÑƒÑ… Ğ´Ğ¾ Ğ±Ğ°Ñ‡ĞµĞ½Ğ¸Ñ… Ñ‚Ğ¾Ñ‡Ğ¾Ğº."""
    new_moves, old_moves = analyze_moves(G, current_pos, visited)
    possible_moves = [node for node in new_moves + old_moves if node not in avoid_nodes]

    # Ğ¯ĞºÑ‰Ğ¾ ĞºÑ–Ğ½Ñ†ĞµĞ²Ğ° Ñ‚Ğ¾Ñ‡ĞºĞ° Ñ” ÑĞµÑ€ĞµĞ´ Ğ±Ğ°Ñ‡ĞµĞ½Ğ¸Ñ… Ğ²ÑƒĞ·Ğ»Ñ–Ğ², Ñ€ÑƒÑ… Ğ´Ğ¾ Ğ½ĞµÑ— Ğ¿Ğ¾ĞºÑ€Ğ¾ĞºĞ¾Ğ²Ğ¸Ğ¹
    if end in seen:
        print(f"ĞšÑ–Ğ½Ñ†ĞµĞ²Ğ° Ñ‚Ğ¾Ñ‡ĞºĞ° {end} Ğ±ÑƒĞ»Ğ° Ğ¿ĞµÑ€ĞµĞ³Ğ»ÑĞ½ÑƒÑ‚Ğ°, Ñ€ÑƒÑ…Ğ°Ñ”Ğ¼Ğ¾ÑÑ Ğ¿Ğ¾ĞºÑ€Ğ¾ĞºĞ¾Ğ²Ğ¾!")
        shortest_path = nx.shortest_path(G, current_pos, end)
        return shortest_path[1], False  # ĞŸĞ¾Ğ²ĞµÑ€Ñ‚Ğ°Ğ½Ğ½Ñ Ğ½Ğ°ÑÑ‚ÑƒĞ¿Ğ½Ğ¾Ğ³Ğ¾ ĞºÑ€Ğ¾ĞºÑƒ Ñƒ ÑˆĞ»ÑÑ…Ñƒ

    if not possible_moves or in_dead_end:
        return _handle_dead_end(G, current_pos, seen, avoid_nodes, in_dead_end)

    return min(possible_moves, key=lambda pos: abs(pos[0] - end[0]) + abs(pos[1] - end[1])), False




def _update_direction(direction, current_pos, next_pos):
    """ĞĞ½Ğ¾Ğ²Ğ»ĞµĞ½Ğ½Ñ Ğ½Ğ°Ğ¿Ñ€ÑĞ¼ĞºÑƒ Ñ€ÑƒÑ…Ñƒ Ğ°Ğ³ĞµĞ½Ñ‚Ğ°."""
    new_direction = get_direction(current_pos, next_pos)
    if new_direction != direction:
        if (direction, new_direction) in [("up", "right"), ("right", "down"), ("down", "left"), ("left", "up")]:
            print("ĞŸĞ¾Ğ²Ğ¾Ñ€Ğ¾Ñ‚ Ğ½Ğ°Ğ¿Ñ€Ğ°Ğ²Ğ¾")
        elif (direction, new_direction) in [("up", "left"), ("left", "down"), ("down", "right"), ("right", "up")]:
            print("ĞŸĞ¾Ğ²Ğ¾Ñ€Ğ¾Ñ‚ Ğ½Ğ°Ğ»Ñ–Ğ²Ğ¾")
        else:
            print("Ğ Ğ¾Ğ·Ğ²Ğ¾Ñ€Ğ¾Ñ‚ Ğ½Ğ° 180 Ğ³Ñ€Ğ°Ğ´ÑƒÑÑ–Ğ²")
    return new_direction


def _finalize_run(current_pos, end, kb, path, visited, step_count):
    """ĞÑÑ‚Ğ°Ñ‚Ğ¾Ñ‡Ğ½Ğµ Ğ¾Ğ½Ğ¾Ğ²Ğ»ĞµĞ½Ğ½Ñ Ğ¿Ñ€Ğ¸ Ğ´Ğ¾ÑÑĞ³Ğ½ĞµĞ½Ğ½Ñ– Ñ†Ñ–Ğ»Ñ– Ğ°Ğ±Ğ¾ Ğ·Ğ°Ğ²ĞµÑ€ÑˆĞµĞ½Ğ½Ñ– Ğ·Ğ° ĞºÑ–Ğ»ÑŒĞºÑ–ÑÑ‚Ñ ĞºÑ€Ğ¾ĞºÑ–Ğ²."""
    if current_pos == end:
        kb.tell_visited(current_pos)
        path.append(current_pos)
        visited.add(current_pos)
        print(f"ğŸ Ğ”Ğ¾ÑÑĞ³Ğ½ÑƒÑ‚Ğ¾ ĞºÑ–Ğ½Ñ†ĞµĞ²Ğ¾Ñ— Ñ‚Ğ¾Ñ‡ĞºĞ¸! Ğ”Ğ¾Ğ´Ğ°Ñ {current_pos} Ğ´Ğ¾ visited_nodes.")

    print("âœ… Ğ ÑƒÑ… Ğ·Ğ°Ğ²ĞµÑ€ÑˆĞµĞ½Ğ¾!")
    print(f"ğŸ“Š Ğ¤Ğ°ĞºÑ‚Ğ¸Ñ‡Ğ½Ğ° ĞºÑ–Ğ»ÑŒĞºÑ–ÑÑ‚ÑŒ ĞºÑ€Ğ¾ĞºÑ–Ğ² Ğ°Ğ³ĞµĞ½Ñ‚Ğ°: {step_count}")





# ğŸš€ğŸš€ğŸš€ğŸš€ Ğ—Ğ°Ğ¿ÑƒÑĞº Ñ‚Ğ° Ñ‚ĞµÑÑ‚ÑƒĞ²Ğ°Ğ½Ğ½Ñ ğŸš€ğŸš€ğŸš€ğŸš€

def test_agent_behavior():
    """Ğ¢ĞµÑÑ‚ÑƒĞ²Ğ°Ğ½Ğ½Ñ Ñ€Ğ¾Ğ±Ğ¾Ñ‚Ğ¸ Ğ°Ğ³ĞµĞ½Ñ‚Ğ° Ğ· Ğ³Ñ€Ğ°Ñ„Ğ¾Ğ¼."""
    test_graph_width, test_graph_height = 5, 5
    test_start, test_end = (0, 0), (4, 4)

    G = nx.grid_2d_graph(test_graph_width, test_graph_height)

    # Ğ’Ğ¸Ğ´Ğ°Ğ»ĞµĞ½Ğ½Ñ Ğ²Ğ¸Ğ¿Ğ°Ğ´ĞºĞ¾Ğ²Ğ¸Ñ… 5 Ñ€ĞµĞ±ĞµÑ€ Ñ–Ğ· Ğ³Ñ€Ğ°Ñ„Ğ°
    G = remove_random_edges(G, remove_count=25)

    generate_speed_limit_signs(G)

    # Ğ±Ğ°Ğ·Ğ° Ğ·Ğ½Ğ°Ğ½ÑŒ Ñ–Ğ½Ñ–Ñ†Ñ–Ğ°Ğ»Ñ–Ğ·Ğ¾Ğ²Ğ°Ğ½Ğ° Ğ¿ĞµÑ€ĞµĞ´ Ğ²Ğ¸ĞºĞ»Ğ¸ĞºĞ¾Ğ¼ Ñ„ÑƒĞ½ĞºÑ†Ñ–Ñ— Ñ€ÑƒÑ…Ñƒ Ğ°Ğ³ĞµĞ½Ñ‚Ğ°, Ğ½Ğµ Ñ€ÑƒÑ…Ğ°Ñ‚Ğ¸
    kb = KnowledgeBase()

    assert nx.has_path(G, test_start, test_end), "Ğ“Ñ€Ğ°Ñ„ Ñ€Ğ¾Ğ·Ñ–Ñ€Ğ²Ğ°Ğ½Ğ¸Ğ¹, Ğ½ĞµĞ¼Ğ°Ñ” ÑˆĞ»ÑÑ…Ñƒ Ğ¼Ñ–Ğ¶ ÑÑ‚Ğ°Ñ€Ñ‚Ğ¾Ğ¼ Ñ– Ñ„Ñ–Ğ½Ñ–ÑˆĞµĞ¼!"

    visited_nodes, path_taken, seen_nodes, speed_limits, total_steps = move_agent_with_knowledge(G, kb, test_start, test_end, max_steps=100)

    print(f"Ğ¨Ğ»ÑÑ… Ğ°Ğ³ĞµĞ½Ñ‚Ğ°: {path_taken}")
    print(f"ĞĞ±Ğ¼ĞµĞ¶ĞµĞ½Ğ½Ñ ÑˆĞ²Ğ¸Ğ´ĞºĞ¾ÑÑ‚Ñ– Ğ½Ğ° Ğ¼Ğ°Ñ€ÑˆÑ€ÑƒÑ‚Ñ–: {speed_limits}")
    print(f"Ğ¤Ñ–Ğ½Ğ°Ğ»ÑŒĞ½Ğ° Ğ¿Ğ¾Ğ·Ğ¸Ñ†Ñ–Ñ: {path_taken[-1]}")


    
    assert test_end in visited_nodes, "ĞĞ³ĞµĞ½Ñ‚ Ğ½Ğµ Ğ´Ğ¾ÑÑĞ³Ğ½ÑƒĞ² Ñ†Ñ–Ğ»ÑŒĞ¾Ğ²Ğ¾Ñ— Ñ‚Ğ¾Ñ‡ĞºĞ¸!"
    assert total_steps > 0, "ĞĞ³ĞµĞ½Ñ‚ Ğ½Ğµ Ğ·Ñ€Ğ¾Ğ±Ğ¸Ğ² Ğ¶Ğ¾Ğ´Ğ½Ğ¾Ğ³Ğ¾ ĞºÑ€Ğ¾ĞºÑƒ!"
    assert len(visited_nodes) > 0, "Ğ–Ğ¾Ğ´ĞµĞ½ Ğ²ÑƒĞ·Ğ¾Ğ» Ğ½Ğµ Ğ±ÑƒĞ² Ğ²Ñ–Ğ´Ğ²Ñ–Ğ´Ğ°Ğ½Ğ¸Ğ¹!"
    assert len(speed_limits) > 0, "Ğ¨Ğ²Ğ¸Ğ´ĞºÑ–ÑÑ‚ÑŒ Ğ´Ğ»Ñ Ğ²ÑƒĞ·Ğ»Ñ–Ğ² Ğ½Ğµ Ğ±ÑƒĞ»Ğ° Ğ¾Ñ‚Ñ€Ğ¸Ğ¼Ğ°Ğ½Ğ°!"

    
    print("âœ… Ğ£ÑÑ– Ñ‚ĞµÑÑ‚Ğ¸ Ğ¿Ñ€Ğ¾Ğ¹Ğ´ĞµĞ½Ñ– ÑƒÑĞ¿Ñ–ÑˆĞ½Ğ¾!")

# Ğ’Ğ¸ĞºĞ¾Ğ½Ğ°Ğ½Ğ½Ñ Ñ‚ĞµÑÑ‚Ñ–Ğ²


    
test_agent_behavior()
```
